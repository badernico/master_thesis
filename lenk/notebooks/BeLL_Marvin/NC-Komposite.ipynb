```python
def array_to_256(array):
    return (array*255.999).astype("uint8")

def day_natural_composite(vis006_data,vis008_data,nir016_data,factor=1,gamma=1):
    blue = array_to_256(np.clip(vis006_data/factor,0,1)**(1./gamma))
    green = array_to_256(np.clip(vis008_data/factor,0,1)**(1./gamma))
    red = array_to_256(np.clip(nir016_data/factor,0,1)**(1./gamma))
    
    return np.dstack([red,green,blue]).astype("uint8")

def time_index2time(ind,dt=5):
    time = ind*dt
    
    h,m = divmod(ind*dt,60)
    
    return (h,m)
```

```python
nc_composites = []

for sf in tqdm.tqdm(sorted(sat_files)):
    sd = xr.open_dataset(sf)
    
    sat_data = {'VIS006':[],'VIS008':[],'IR_016':[],'IR_108':[]}
    
    for k in sat_data.keys():
        sat_data[k].append(sd[k].data)
        
    nc_composites.append(day_natural_composite(sat_data['VIS006'][0][0],
                                               sat_data['VIS008'][0][0],
                                               sat_data['IR_016'][0][0],
                                               0.9,1.8))    
    sd.close()
```

```python
from matplotlib.image import imread
from tempfile import NamedTemporaryFile

def get_size(fig, dpi=100):
    with NamedTemporaryFile(suffix='.png') as f:
        fig.savefig(f.name, bbox_inches='tight', dpi=dpi)
        height, width, _channels = imread(f.name).shape
        return width / dpi, height / dpi

def set_size(fig, size, dpi=100, eps=1e-2, give_up=2, min_size_px=10):
    target_width, target_height = size
    set_width, set_height = target_width, target_height # reasonable starting point
    deltas = [] # how far we have
    while True:
        fig.set_size_inches([set_width, set_height])
        actual_width, actual_height = get_size(fig, dpi=dpi)
        set_width *= target_width / actual_width
        set_height *= target_height / actual_height
        deltas.append(abs(actual_width - target_width) + abs(actual_height - target_height))
        if deltas[-1] < eps:
            return True
        if len(deltas) > give_up and sorted(deltas[-give_up:]) == deltas[-give_up:]:
            return False
        if set_width * dpi < min_size_px or set_height * dpi < min_size_px:
            return False
```

```python
slon,slat = mst.get_msg_lon_lat('eu')
```

```python
(x,y),m = bmaps.make_map(slon,slat,region='meu',color='lightblue')
```

```python
def plot_ir108_and_nc(satellite_data,time_string,mp=m,x=x,y=y):
    nc = day_natural_composite(satellite_data['VIS006'].data[0],
                               satellite_data['VIS008'].data[0],
                               satellite_data['IR_016'].data[0],0.9,1.8)
    
    color_tuples = np.array([nc[:-1,:-1,0].flatten()/255.999,
                         nc[:-1,:-1,1].flatten()/255.999,
                         nc[:-1,:-1,2].flatten()/255.999]).transpose()
    
    title_ir = r"$T_\mathrm{B}$(10,8 ¬µm)"
    fig,ax = plt.subplots(1,2,figsize=(20/2.54,10/2.54),sharex=True,sharey=True)
    plt.sca(ax[0])
    p = mp.pcolormesh(x,y,
                     satellite_data['IR_108'].data[0],
                     vmin=210,vmax=300,cmap=emap)
    (x,y),mp = bmaps.make_map(slon,slat,region='meu',mp=mp,color='purple',lw=1)
    cbar = fc.colourbar(p)
    cbar.set_label(r'{} / K'.format(title_ir), rotation=90)
    ax[0].set_title(r"{}, 24.05.2010 {} UTC".format(title_ir,time_string))

    plt.sca(ax[1])
    im = ax[1].pcolormesh(x,y, np.zeros((599,799)), color=color_tuples)
    im.set_array(None)
    (x,y), mp = bmaps.make_map(slon,slat,region='meu',mp=mp,color='black',lw=1)
    ax[1].set_title("Naturfarbenkomposit, 24.05.2010 {} UTC".format(time_string))
    
    plt.tight_layout()
    plt.savefig("{}/2019-09_bell/images/IR_NC/IR_NC_20150524T{}Z.png".format(proj_path,time_string.replace(":","")),bbox_inches="tight")
    plt.close("all")
```

```python
for sf in tqdm.tqdm(sorted(sat_files)):
    sd = xr.open_dataset(sf)
    
    time_string = sf.split("/")[-1].split("_")[2].split("-")[0].split("T")[1].split("Z")[0]
    
    plot_ir108_and_nc(sd,"{}:{}".format(time_string[:2],time_string[-2:]))
    sd.close()
```
