```python
import numpy as np
import datetime as dt

import matplotlib.pyplot as plt
%matplotlib inline

import read_linet_data as rld
from io_tools import radolan

from plotting_tools.bmaps import radolan_rx_map
from scipy.ndimage import filters
```

```python
r = radolan.Radolan()
```

```python
lightnings = rld.read_linet_for_time_intervall(dt.datetime(2010,5,24,11,50),dt.datetime(2010,5,25,15,15))
```

```python
lighning_grid = rld.linet2rado_frequency_grid(lightnings)
```

```python
m = radolan_rx_map()
```

```python
fig,ax,m = radolan_rx_map(linecolour='black',figsize=(14,14))
x,y = m(r.lon,r.lat)
plt.sca(ax)
p = ax.pcolormesh(x,y,filters.gaussian_filter(lighning_grid[0],1),cmap='gist_heat_r')
plt.colorbar(p)
```

```python
fig,ax = plt.subplots(1,1,figsize=(16,16))
p = ax.pcolormesh(r.lon,r.lat,lighning_grid[2],vmin=-100,vmax=100,cmap='RdBu_r')
plt.colorbar(p)
```
