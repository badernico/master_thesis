# Edge in Radar Field

```python
%matplotlib inline

import pylab as pl
import numpy as np
import datetime
import io_tools.radolan as rado
import scipy.ndimage
```

## Get Data

```python
t = datetime.datetime(2012, 7, 28, 11, 0)

r = rado.Radolan()
r.read(t)
```

```python
pl.pcolormesh(r.lon, r.lat, r.data)
```

```python
pl.pcolormesh(r.lon, r.lat, r.data.mask)
outside_mask = r.data.mask
```

## Use Dilatation to increase Edge Mask

```python
npix = 11
extended_outside_mask = scipy.ndimage.maximum_filter(outside_mask, npix)

pl.pcolormesh(r.lon, r.lat, extended_outside_mask)

```

## Edge Extraction

```python
edge_mask = extended_outside_mask & (outside_mask == False)

pl.pcolormesh(r.lon, r.lat, edge_mask)
```

## Include Domain Edges

```python
edge_mask[0:npix] = True
edge_mask[:,0:npix] = True
edge_mask[-npix:] = True
edge_mask[:,-npix:] = True

pl.pcolormesh(r.lon, r.lat, edge_mask)
```

## Combine RX data with edge

```python
thresh = 30.

rx_mask = r.data >= 35

rx_mask = rx_mask | edge_mask
pl.pcolormesh(r.lon, r.lat, rx_mask)


```
