## Trackverläufe

Jetzt laden wir die Dateinamen aller Tracks.

```python
tracks = glob.glob("/vols/talos/home/stephan/data/life_cycle/Handtracks/trackdata/*.nc")
tracks.sort()
```

Wir erstellen eine Liste mit den Interessenfeldervariablennamen, um sie dann zu durchlaufen.

```python
felder = ['b108',
    'b62_73',
    'indicator_sum',
    'cool_30',
    'cool_15',
    'b87_108',
    'db62_73_t15',
    'dtri_t15',
    'b62_108',
    'db62_73_t30',
    'db87_108_t15']
```

Nun testen wir die Darstellung erstmal mit einem Track.

```python
tr = tracks[273]
daten = trackdaten_laden.trackdaten_laden(tr)
rel_zeit = sort_dtbmin(daten,5)

tb180 = [t[25,25] for t in daten['b108']]
print tb180

plt.plot(rel_zeit,tb180)
```

Das scheint ja zu funktionieren. Deshalb nun einen Versuch mit dem Median aus einem 3x3-Pixelausschnitt.

```python
tb180_stat = feldstatistik(daten['b108'],3)
plt.plot(rel_zeit,tb180_stat['med'])
```

Das sieht auch vernünftig aus. Dann können wir das ja mal für alle Tracks versuchen.

```python
fig = plt.figure(figsize=(10,10))
sns.set_style('whitegrid')
for tr in tracks:
    daten = trackdaten_laden.trackdaten_laden(tr)
    rel_zeit = sort_dtbmin(daten,5)
    tb180_stat = feldstatistik(daten['b108'],3)
    try:
        plt.plot(rel_zeit,tb180_stat['med'],'b-',alpha=0.3)
    except:
        nachricht = "Fehler bei Track %s" %tr
        print(nachricht)
        continue
```

Und die Darstellung nochmal für die Maxima.

```python
%config InlineBackend.figure_formats=['svg']

for tr in tracks:
    daten = trackdaten_laden.trackdaten_laden(tr)
    rel_zeit = sort_dtbmin(daten,5)
    tb180_stat = feldstatistik(daten['b108'],3)
    try:
        plt.plot(rel_zeit,tb180_stat['max'],'b-',alpha=0.3)
    except:
        nachricht = "Fehler bei Track %s" %tr
        print(nachricht)
        continue
```
