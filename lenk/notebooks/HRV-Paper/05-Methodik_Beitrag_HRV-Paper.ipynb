##### Adaptiver Schwellwert

Hartwigs Vorschlag sieht vor, die Objekte wie folgt zu definieren:

S = max(T<sub>min</sub>&thinsp;+&thinsp;&sigma;, S<sub>max.</sub>)

mit:
- S: Schwellwert zur Objektmaskierung
- T<sub>min</sub>: Minimum (lokal oder global)
- &sigma;: Spreizungsfaktor vom Minimum aus, um Objektrand zu definieren
- S<sub>max.</sub>: Maximalwert für den Objektrand

```python
def get_mask_threshold(max_threshold,spread,field=None,object_min=240,):
    """
    Determine threshold from given field or object minimum. If the minimum is lower than the given threshold,
    the threshold to determine the cloud edge is determined using the given spread.
   
    Inputs:
        * max_threshold: int or float
            maximum threshold of the cloud edge
        * spread: int or float
            spread to determine cloud edge, if the minimum is lower than the threshold
        * field: array-like, default=None
            numpy array with the field data
        * object_min: int or float, default=240
            object minimum
    """
        
    if np.all(field) != None:
        # get minimum of the field
        minimum = np.min(field)
    else:
        minimum = object_min
   
    # return threshold
    return np.max((minimum+spread,max_threshold))    
```

Dieser Ansatz ist für lokale Minima sinnvoll, aber für globale Minima wird, in Abhängigkeit vom maximalen Schwellwert und Spreizungsfaktor, für die meisten Zeitschritte der maximale Schwellwert als Maskierungsschwellwert gewählt werden. Erst ab einem maximalen Schwellwert von 230&thinsp;K sollte sich der Schwellwert adaptiv verhalten.

```python
thresh = np.arange(270,210,-10)
spreads = np.arange(0,11,1)
```

```python
thresholds = {t:{s:[] for s in spreads} for t in thresh}
for t in thresh:
    for s in spreads:
        for ir in ir108_subset:
            max_thresh = get_mask_threshold(t,s,ir)
            
            thresholds[t][s].append(max_thresh)
```

```python
norm = mpl.colors.Normalize(vmin=1, vmax=10)
```

```python
linestyles = {270:(0,()),
              260:(0,(5,1)),
              250:(0,(5,2)),
              240:(0,(5,3)),
              230:(0,(5,4)),
              220:(0,(5,5))}
```

```python
linelabel = ["{} K".format(t) for t in thresh]
spread_label = ["{} K".format(s) for s in np.arange(0,11,1)]
```

```python
fig,ax = plt.subplots(1,1,figsize=(14,7))
for i,t in enumerate(thresholds.keys()):
    for j,s in enumerate(thresholds[t].keys()):
        p0 = ax.plot(time_subset,
                     thresholds[t][s],
                     c=mpl.cm.nipy_spectral(norm(j)),
                     linestyle=linestyles[t],
                     alpha=0.7)   

# resize plot
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])

# create legends
dummy_lines = []
for t_idx, t in enumerate(thresholds.keys()):
    dummy_lines.append(ax.plot([],[], c="black", ls = linestyles[t])[0])
lines = ax.get_lines()

legend1 = plt.legend([lines[i] for i in spreads], 
                     spread_label, 
                     loc="center left",
                     title="Spread",
                     bbox_to_anchor=(1, 0.5))

legend2 = plt.legend([dummy_lines[i] for i,t in enumerate(thresholds.keys())],
                     linelabel,
                     loc="upper center",
                     title="Threshold",
                     bbox_to_anchor=(0.5, -0.08),
                     ncol=6)
ax.add_artist(legend1)

ax.xaxis.set_major_formatter(hourminuteFmt)
ax.set_xlim(dt.datetime(2013,6,18,13,25),dt.datetime(2013,6,18,14,15))
ax.set_xlabel("time / UTC")
ax.set_ylabel("scene threshold / K") 
```

#### 2.1.3 Glättung des <i>T</i><sub>B</sub>&thinsp;10,8&thinsp;µm-Feldes vor der Minimabestimmung

Das Feld vor der Bestimmung der Minima mit einem 2D-Gaußfilter zu glätten sollte das Problem mit mehren lokalen Minima für eine Zelle hoffentlich lösen können. Um die Beispielzelle in den früheren Zeitschritten nicht zu verlieren, sollte die Glättung eher schwach sein.

```python
gauss_sigmas = np.arange(0,2,0.1)
```

```python
ir_smoothed = {gs:[] for gs in gauss_sigmas}
lm_smoothed = {gs:[] for gs in gauss_sigmas}
```

```python
for gs in gauss_sigmas:
    smoothed = ndi.gaussian_filter(ir108_subset,gs)
    
    ir_smoothed[gs] = smoothed
    
    lmin = [h_minima(sm,10) for sm in smoothed]
    
    lm_l = [ndi.label(lm)[0] for lm in lmin]
    
    lm_smoothed[gs] = lm_l
```

### 2.2 Objektmaskierung

Als nächstes definieren wir die Objekte.
   
#### 2.2.1 Objektschwellwerte

Ausgehend von den lokalen Minima betimmen wir jetzt die Objektränder nach Hartwigs Ansatz. Da die Beispielzelle im ersten Zeitschritt noch relativ warm ist, und wir zuviele Randpixel vermeiden wollen, müssen wir den Ansatz wohl etwas modifizieren:

1. Als maximalen Schwellwert setzen wir 270&thinsp;K, um das Beispielobjekt im ersten Zeitschritt definieren zu können.

2. Wenn das Objektminimum kleiner als dieser Schwellwert ist, wird der Schwellwert für die Maskierung aus dem Objektminimum&thinsp;+&thinsp;den Streuungsfaktor bestimmt.

3. Der Streuungsfaktor wird als Minimum aus der Differenz von 270&thinsp;K und dem Objektminimum oder dem vorgegebenen Streuungsfaktor bestimmt.
4. Der Maskierungssschwellwert S ist:
   S = min(<i>T</i><sub>min</sub>&thinsp;+&thinsp;&sigma;, S<sub>max</sub>)

```python
def get_mask_threshold_mod(field,max_threshold,spread):
    """
    Determine threshold from given field. If the field minimum is lower than the given threshold,
    the threshold to determine the cloud edge is determined using the given spread.
    
    Inputs:
        * field: array-like
            numpy array with the field data
        * threshold: int or float
            threshold of the cloud edge
        * spread: int or float
            spread to determine cloud edge, if the minimum is lower than the threshold
    """
    # get minimum of the field
    field_min = np.min(field)
    
    # determine spread
    spread = np.min((max_threshold-field_min,spread))
    
    # return threshold
    return np.min((field_min+spread,max_threshold))
```

```python
lm_values = {t:{} for t in np.arange(0,9)}
lm_thresholds = {t:{} for t in np.arange(0,9)}

for i,lm in enumerate(lm_smoothed[0.2]):
    minima_ids = np.unique(lm)[1:]
    
    for m in minima_ids:
        min_location = np.where(lm==m)
        min_value = np.unique(ir_smoothed[0.2][i][min_location])
        
        thresh = get_mask_threshold_mod(np.array([min_value]),271,10)
        lm_thresholds[i][m] = thresh
        lm_values[i][m] = min_value[0]
```

#### 2.2.2 Segmentierung
Als nächstes folgt die Maskierung der Objekte mit ihrem jeweiligen Maskierungsschwellwert. 

Dazu müssen wir ausgehend vom lokalen Minimum das Objekt bis zu dem Maskierungsschwellwert konstruieren. Das geht wahrscheinlich gut mit dem Watershed-Algorithmus, aber vielleicht geht es auch einfacher mit einer Maskierung.

##### Maskierung

Die Maskierung erfolgt folgendermaßen:
1. Wir durchlaufen alle lokalen Minima für eine Szene und sammeln ihre Werte und die dazugehörigen Werte für die Objektgrenzen.

2. Wir maskieren für jedes lokale Minimum aller Werte in der die Szene, die größer als der Wert für die Objektgrenzen sind.

3. Da wir jetzt oftmals mehrere mögliche Objekte haben, und die Maske eingetlich nur für eines gilt, erzeugen wir Objektkennzeichnungen.

4. Wir behalten nur das Objekt, in dem sich das lokale Minimum befindet.

5. Wir wiederholen das für alle lokalen Minima und Zeitschritte.

```python
obj_labeled = []

for i, ir in enumerate(ir_smoothed[0.2]):
    labels = np.unique(lm_smoothed[0.2][i])[1:]
    
    segmented = np.zeros_like(ir)
    
    masks = []
    
    for l in labels:
        mask = np.ma.masked_greater(ir,lm_thresholds[i][l])
        mask_labeled = ndi.label(~mask.mask*1)[0]
        lokmin_location = np.where(lm_smoothed[0.2][i] == l)
        wanted_obj_id = mask_labeled[lokmin_location]
        wanted_obj_location = np.where(mask_labeled==wanted_obj_id)
        segmented[wanted_obj_location] = l
        
        masks.append(masks)
    
    obj_labeled.append(segmented)
```
