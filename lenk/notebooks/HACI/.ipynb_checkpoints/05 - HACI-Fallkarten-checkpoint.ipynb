## 1. Daten laden

```python
haci_cases = pd.read_csv("{}/HACI_bbox_data/HACI_2008-2017-bbox-filtered.csv".format(local_data_path))
```

```python
haci_cases = haci_cases[haci_cases.darea>10]
```

```python
hc = pd.read_csv("{}/HACI_bbox_data/haci-2008-2017-bbox.dat".format(local_data_path),sep=",")
```

```python
hc.iloc[0].l0
```

```python
hc_tf = hc[hc.dt >= 6]
```

```python
len(hc_tf.dt)
```

```python
l = hc_tf.l0 + (hc_tf.l1 - hc_tf.l0)//2
c = hc_tf.c0 + (hc_tf.c1 - hc_tf.c0)//2
```

```python
l = l.values
c = c.values
```

```python
l
```

```python
points = np.array((np.rint(l).astype("int16"),
                   np.rint(c).astype("int16"))).T
points = list(map(tuple,points))
```

```python
def count(listOfTuple): 
    """
    Counts occurences of a tuple.
    """
    import collections 

    val = collections.Counter(listOfTuple) 
   # uniqueList = list(set(listOfTuple)) 
      
    return val
```

```python
# count occureces of tuples
counts = count(points)
```

```python
counts
```

```python
haci_grid_tf = np.zeros((900,900))
```

```python
for c in counts:
    if c[0] > 899 or c[1] > 899 or c[0]<0 or c[1]<0:
        if verbose == True:
            print('\x1b[1;30;43m' + "Lightning outside of Radolan domain." + '\x1b[0m')
        else:
            continue
    else:
        haci_grid_tf[c[0],c[1]] = counts[c]      
```

```python
haci_grid_f = np.zeros((900,900))
lf = haci_cases.l0 + (haci_cases.l1 - haci_cases.l0)//2
cf = haci_cases.c0 + (haci_cases.c1 - haci_cases.c0)//2
```

```python
lf = lf.values
cf = cf.values
```

```python
pointsf = np.array((np.rint(lf).astype("int16"),
                   np.rint(cf).astype("int16"))).T
pointsf = list(map(tuple,pointsf))
```

```python
countsf = count(pointsf)
```

```python
for c in countsf:
    if c[0] > 899 or c[1] > 899 or c[0]<0 or c[1]<0:
        if verbose == True:
            print('\x1b[1;30;43m' + "Lightning outside of Radolan domain." + '\x1b[0m')
        else:
            continue
    else:
        haci_grid_f[c[0],c[1]] = countsf[c]  
```

```python
z_grid= np.zeros((900,900))
lz = hc.l0 + (hc.l1 - hc.l0)//2
cz = hc.c0 + (hc.c1 - hc.c0)//2
lz = lz.values
cz = cz.values
```

```python
points_z = np.array((np.rint(lz).astype("int16"),
                     np.rint(cz).astype("int16"))).T
points_z = list(map(tuple,points_z))
```

```python
counts_z = count(points_z)
for c in counts_z:
    if c[0] > 899 or c[1] > 899 or c[0]<0 or c[1]<0:
        if verbose == True:
            print('\x1b[1;30;43m' + "Lightning outside of Radolan domain." + '\x1b[0m')
        else:
            continue
    else:
        z_grid[c[0],c[1]] = counts_z[c]  
```

```python
r = radolan.Radolan()
```

```python
x,y = rmap(r.lon,r.lat)
```

```python
haci_grid_resampled = transform.downscale_local_mean(haci_grid_tf, (10,10))
```

```python
fig,ax,rmap = radolan_rx_map(rmap,linecolour='k')
p = rmap.pcolormesh(x,y,z_grid,norm=mcolors.LogNorm(),cmap='gist_heat_r',vmin=0.1)
cbar = fc.colourbar(p)
cbar.set_label(u'Anzahl / kmÂ²', rotation=90)
ax.set_title(r"Anzahl von Z $\geq$ 35 dBZ, 2008 - 2017")
ax.text(0.05, 0.95, "N = {}".format(len(hc.l0)), horizontalalignment='left',
        verticalalignment='center', transform=ax.transAxes,
        bbox=dict(boxstyle="round",
                   ec=(0, 0, 0),
                   fc=(1, 1, 1),
                   )
         )
plt.savefig("{}/2020-05_EMF-Vortrag/Z_2008-2017.png".format(proj_path),bbox_inches="tight")
```
