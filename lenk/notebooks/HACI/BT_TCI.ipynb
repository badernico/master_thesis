```python
import numpy as np
import glob
import xarray as xr
import tqdm
import pandas as pd
```

```python
path = "/vols/satellite/home/lenk/data/HACI_track_data"
```

```python
track_files = glob.glob("{}/*.nc".format(path))
```

```python
track_files
```

```python
track_files[0].split("/")[-1].split(".")[0][6:]
```

```python
mins = dict()
```

```python
for tf in tqdm.tqdm(track_files):
    tid = tf.split("/")[-1].split(".")[0][6:]
    try:
        with xr.open_dataset(tf) as file:
            mins[tid] = np.min(file.ir108.data)
    except:
        continue
```

```python
mins
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_context("talk")
```

```python
fig,ax = plt.subplots(1,1,figsize=(15,7))
n0, bins0, patches0 = ax.hist(mins.values(),bins=70,density=True,color='#0011ff33',edgecolor='b')
ax.set_ylabel('relative frequency', color='blue') 
ax.tick_params(axis='y', labelcolor='blue')
ax.set_xlabel("10.8 µm brightness temperature / K")
ax.text(260,0.025,"time frame: 2012 to 2014")
ax.text(260,0.0225,"N = {}".format(len(mins.values())))
        
ax1 = ax.twinx()  # instantiate a second axes that shares the same x-axis
color = 'red'
n1, bins1, patches1 = ax1.hist(mins.values(), 70, density=True, histtype='step',
                               cumulative=True, label='Empirical',color='red',linewidth=2)
ax1.set_ylabel('cumulative frequency', color=color)  # we already handled the x-label with ax1
ax1.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.xlim(199,289)
plt.title("Minimum BT(10.8 µm) of 51x51px MSG-Pixel cutout around HACI-centroid at CI time, 2012 - 2014")
plt.savefig("/vols/satellite/home/lenk/notebooks/HACI/ir108_tci.pdf")
```
