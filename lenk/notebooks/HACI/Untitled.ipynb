```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def local_threshold(T, Tmax=273.15, thresh_min=240, Tmin=220, spread=5, factor = 10):
    T_thresh = np.max((T + spread, T + spread + factor*(T - Tmax)/ (Tmin - Tmax),thresh_min))
    return T_thresh
```

```python
spread = 7
factor = [5,10,15]
Tmax = 273.15
Tmin = 220
```

```python
T = np.arange(300,190,-1)
```

```python
f0 = np.array([t + spread for t in T])
f1 = np.array([local_threshold(t,Tmax,240,220,7,factor[0]) for t in T])
f2 = np.array([local_threshold(t,Tmax,240,220,7,factor[1]) for t in T])
f3 = np.array([local_threshold(t,Tmax,240,220,7,factor[2]) for t in T])
```

```python
fig,ax = plt.subplots(1,1,figsize=(16,10))
ax.plot(T,T,'k--')
ax.plot(T,f1,label="5K")
ax.plot(T,f2,label="10K")
ax.plot(T,f3,label="15K")
ax.legend()
ax.axvline(273.15,color='k',linestyle=":")
```
