# Optimierung von mathematischen Graphen

Nach dem Tutorial auf https://www.datacamp.com/community/tutorials/networkx-python-graph-tutorial

Graphen sind Strukturen, die die Beziehungen zwischen Objekten abbilden. Die Objekte werden als <b>Knoten</b> und die Verbindungen als <b>Kanten</b> bezeichnet.

Der Ausgangssgraph ist <b>ungerichtet</b>, d.h. die Kanten sind unorientiert, also <b>biderektional</b>, z.B. A &#8596; B &equiv; B &#8596; A. Ein Graph, der einem die kürzeste Entfernung zu einem Ort anggibt, kann aber auch <b>gerichtet</b> sein, d.h. die Reihenfolge und die Richtung spielen eine Rolle, z.B. A &#8594; B &ne; B  &#8594; A.

Die Kanten eines Graphen können mit Attributen versehen werden. So kann ein Graph <b>kantengewichtet</b> sein, wenn z.B. die Entfernung zwischen den Knoten durch ein Attribut 
als Kantebreite angegeben wird. 

Der <b>Grad</b> eines Graphen beschreibt die Anzahl von Kanten die in einen (sie berührenden) Knoten münden. Knoten werden als <b>ungeradegradige</b> Knoten beschrieben, wenn die Anzahl an Kanten ungerade ist und als <b>geradgradig</b>, wenn sie geradzahlig ist.

Ein <b>Eulerkreis</b> (auch Eulerzug, Eulertour oder Eulersche Linie) ist ein Graph, in dem ein Kreis so definiert werden kann, dass er von einem Startknoten aus exakt einmal durch alle anderen Knoten geht und wieder beim Startknoten ankommt.

Eine <b>Paarung</b> ist eine Untermenge von Kanten, in der jeder Knoten nicht mehr als einmal vorkommt. Ein <b>Paarung mit minimaler Gewichtung</b> findet die <b>Paarung</b> mit der 
kleinsten möglichen Kantengewichtssumme.

## Das Chinesische-Briefträger-Problem

Ein Ansatz um die kürzeste Verbindung zwischen allen Knoten zu finden ist das Chinesische-Briefträger-Problem (CBP). Dieser Algorithmus kann konzeptionell relativ einfach gelöst werden:

1. Fine alle ungeradgradigen Knoten (sehr einfach): Finde alle Pfadüberschneidungen, wo die Anzahl der die Überscheidung berührenden Pfade ungeradzahlig ist.

2. Füge zu dem Graphen Kanten hinzu, so dass alle ungeradgradigen Knoten geradgradig werden. Diese hinzugefügten Kanten müssen Duplikate aus dem Originalgraphen sein (wir nehmen an, dass sich niemand durch die Büsche schlägt). Dieser Kantensatz sollte sich zur kleinsten möglichen Entfernung aufsummieren. (schwer...np-schwer, um genau zu sein): Einfacher gesagt: minimiere die Azahl der Umkehrmöglichkeiten auf einer Route, die jeden Pfad trifft.

3. Finde einen Eulerkreis von einem Startpunkt aus durch dem erweiterten Datensatz (mittelleicht): Wenn wir einmal wissen, auf welchen Pfaden wir umkehren, können wir eigentlich die Route vom Anfang bis zum Ende berechnen.

### Annahmen und Vereinfachungen
#### 1. Nur erforderliche Pfade

Wie aus der Wanderkarte ersichtlich wird, gibt es entlang der Parkgrenzen Straßen, die als Verbindungen zwischen den Pfaden fungieren könnten, insbesonderen der roten Pfade. Es gibt auch einige Pfade (Hufeisenpfade und unmarkierte Farben), die nicht für das Giantmasterlog erforderlich sind, aber nützlich sein könnten um ausgedehnte Umkerhmöglichkeiten zu vermeiden. Die optionalen Pfade einzubeziehen ist eigentlich eine etablierte Variante des CBP, die das Landbriefträgerproblem genannt wird. In diesem Turtorial ignorieren wir diese optionalen Pfade und konzentrieren uns nur auf die erforderlichen.

#### 2. Anname bergauf = bergab

Der CBA nimmt an, dass der Aufwand einen Pfad entlang zu laufen gleicht seiner Entfernugn ist, egal in welcher Richtung man ihn entlang geht. Einige Pfade sind allerdings ziemlich hügelig und erfordern mehr Energie, wenn man sie hinauf geht als hinab. Eine Messgröße, die sowohl Entfernung als auch den Höhenunterschied über einen gerichteten Graphen berücksichtigt, könnte in einer Erweiterung des CBA benutzt werden, die als das Aufgeblasener-Briefträger-Problem bekannt ist.

#### 3. Keine parallelen Kanten (Pfade)
Während es möglich ist parallele Kanten (mehrfache Wege, die dieselben beiden Knoten verbinden) einzubeziehen, führt dies zu mehr Rechenkomlexität. Zum Glück taucht das hier nur zweimal auf (Blau &#8596; roter Diamant und Blau  &#8596 Turmpfad). Das umgehen wir durch einen kleinen Trick in der Kantenlist: doppelte Knoten werden mit dem dupe-Suffix eingebunden um jeden Pfad zu erfassen während man die Einzigartigkeit der Kanteliste erhält. Die CBP-Implementierung in dem Paket, das der Tutorialersteller geschrieben hat, löst das Problem der parallelen Kanten eleganter.

### 2. Finde paare minimaler Entfernung
Das ist das Kernproblem. Man kann es in fünf Punkte einteilen:

1. Berechne alle möglichen Paare ungeradgradiger Knoten.
2. Berechne den kürzesten Pfad zwischen allen Paaren aus Schritt 1.
3. Erzeuge einen geschlossenen Graphen, der jedes Knotenpaar aus Schritt 1 mit der kürzesten Entfernung aus Schritt 2  verbindet.
4. Berechne eine Paarung mit minmaler Gewichtung des Graphen aus Schritt 3. (Das reduziert sich darauf, zu bestimmen, wie man die ungeraden Knoten so paart, dass die Summe der Entfernung zwischen den paaren so klein wie möglich wird).
5. Erweitere den Originalgraphen mit den kürzesten Pfaden zwischen den Knotenpaaren aus Schritt 4.

#### 2.2 Berechnen den kürzesten Pfad zwischen Knotenpaaren

Das ist der erste Schritt, der wirkliche Rechenleistung benötigt. Zum Glück hat das `networkx`-Paket eine schöne Implementierung des Dijkstraalgorithmus, um den kürzesten Pfad zwischen zwei Knoten zu berechnen. Man wendet diese Funktion auf jedes Knotenpaar an (alle 630).

```python
def get_shortest_paths_distances(graph, pairs, edge_weight_name):
    """Compute shortest distance between each pair of nodes in a graph.  Return a dictionary keyed on node pairs (tuples)."""
    distances = {}
    for pair in pairs:
        distances[pair] = nx.dijkstra_path_length(graph, pair[0], pair[1], weight=edge_weight_name)
    return distances
```

#### 2.3 Erzeuge eine geschlossenen Graphen
Ein geschlossener Graph ist einfach ein Graph, in dem jeder Knoten mit jedem anderen Knote durch eine einzigartige Kante verbunden ist.

Unser Graph sollte 36 Knoten und 630 Kanten mit zugehörigem Kantegewicht (Entfernung) haben.

Die Funktion `create_complete_graph` ist dafür definiert ihn zu berehcnen. Der Parameter `flip_weights` wird benutzt, um die Entfernung in ein Gewichtungsattribut umzuwandeln, wo kleinere Zahlen größeren Entfernungen und große Zahlen kürzeren ensprechen. Das klingt ein wenig unintuitiv, aber es ist für Schritt 2.4 notwendig, wo wir die Paarung mit dem kleinsten Gewicht für den ganzen Graphen berechnen. 

Idealerweise berechnet man die Paarung mit dem kleinsten Gewicht direkt, aber `networkx` implementeirt eine `max_weight_mathing`-Funktion, die Kantengewichte maximiert statt sie zu minimeren. Wir umgehen das, indem wir das Entfernungsattribut negieren (mit -1 multiplizieren). Das stellt sicher, das die Reihenfolge und Skalierung druch die Enfernung erhalten, aber umgekehrt werden.

```python
def create_complete_graph(pair_weights, flip_weights=True):
    """
    Create a completely connected graph using a list of vertex pairs and the shortest path distances between them
    Parameters:
        pair_weights: list[tuple] from the output of get_shortest_paths_distances
        flip_weights: Boolean. Should we negate the edge attribute in pair_weights?
    """
    g = nx.Graph()
    for k, v in pair_weights.items():
        wt_i = - v if flip_weights else v
        g.add_edge(k[0], k[1], attr_dict={'distance': v, 'weight': wt_i})
    return g
```

#### 2.5: Originalgraphen erweitern 

Nun erweitern wir den Originalgraphen mit den Kanten aus der Paarung in Schritt 2.4. Eine einfache Funktion dafür, die auch beachtet, dass diese neuen Kanten vom erweitereten Graphen stammen, wird unten definiert. Wir brauchen diese Information im Schritt 3, wenn wir den eigentlichen Eulerkreis durch den Graphen erzeugen. 

```python
def add_augmenting_path_to_graph(graph, min_weight_pairs):
    """
    Add the min weight matching edges to the original graph
    Parameters:
        graph: NetworkX graph (original graph from trailmap)
        min_weight_pairs: list[tuples] of node pairs from min weight matching
    Returns:
        augmented NetworkX graph
    """

    # We need to make the augmented graph a MultiGraph so we can add parallel edges
    graph_aug = nx.MultiGraph(graph.copy())
    for pair in min_weight_pairs:
        graph_aug.add_edge(pair[0],
                           pair[1],
                           attr_dict={'distance': nx.dijkstra_path_length(graph, pair[0], pair[1]),
                                      'trail': 'augmented'}
                          )
    return graph_aug
```

### 3. Eulerkreis berechnen
Da wir nun den geradgradigen Graphen haben, ist die schwere Optimisierungsarbeit geschafft. Wie Euler berühmterweise 1736 mit seinem Problem der Sieben Brücken in Königsberg postuliert hat, gibt einen Pfad der jede Kante genua einmal besucht, wenn alle Knoten geradgradig sind. Carl Hierholzer hat das später in den 1870er Jahren formell bewiesen.

Es gibt viele Euelrkreise mit der gleichen Entfernung, die konstruiert weden können. Man erreicht 90% des Weges dorthin mit der Funktion `eulerian_circuit` des `networkx`-Paketes. Aber es gibt ein paar Einschränkungen.

Einschränkungen, die wir umgehen können:

1. Der erweiterte Graph könnte (und wird wahrscheinlich) Kanten enthalten, die nicht im Originalgraph existieren. Um den Kreis (ohne sich in die Büsche zu schlagen) zu schließen, muss man diese erweiterten Schritte in den kürzesten Pfad durch die Kanten zerlegen, der tatsächlich existiert. 

2. `eulerian_circuit` gibt nur die Reihenfolge zurück, in der jeder Knoten getroffen wird. Sie gibt nicht die Kantenattribute der Kanten zurück, die benötigt werden, um den Kreis zu schließen. Sie sind aber notwendig, da wir verfolgen müssen, an welchen Kanten bereits entlang gegangen wurde, wenn es mehrere Kanten zwischen Knoten gibt.

Einschränkungen, die wir nicht umgehen können:

3. Um uns Laufarbeit zu ersparen, könnte man die  Bedingung des Eulerkreises schwächen, dass der Kreis am selben Knoten anfängt und aufhört. Ein Eulerpfad (der Allgemeinfall eines Eulerkreises) kann auch gefundne werden, wenn es exakt zwie ungeradgradige Knoten gibt. Das würde die Umkerhmöglichkeiten ein wenig veringern ... vorausgesetzt man könnte von anderen Ende des Parkes zurückfahren. Allerdings unterstützt `networkx` zum Zeitpunkt als diese geschrieben wurde, noch keinen Eulerpfadlagorithmus. Der Code von `eulerian_circuit` ist nicht zu schlecht und könnte angepasst werden, aber in diesem Fall wollen wir es nicht zu kompliziert machen.

Wir beginnen mit dem naiven Eulerkreis.

#### Richtiger Eulerkreis
Wir definieren eine Funktion, die den Originalgraphen benutzt, um uns zu sagen welche Pfade genutzt werden sollen, um von Knoten A zu Knoten B zu gelangen. Obwohl der Code ziemlich umfangreich ist, ist die Logik eingentlich recht einfach. Man formt einfach den naiven Kreis, der Kanten beinhaltete, die im Originalgraphen nicht existieren, in einen Graphen um, der nur Kanten beinhaltet, die im Originalgraphen vorhanden sind. 

Man durchlauft alle Kanten in naiven Eulerkreis. Immer, wenn man eine Kante trifft, die nicht im Originalgraphen vorhnden ist, ersetzt man sie mit einer Reihe von Kanten, bestehend aus dem kürzesten Pfad zwischen ihren Knoten im Originalgraphen.

```python
def create_eulerian_circuit(graph_augmented, graph_original, starting_node=None):
    """Create the eulerian path using only edges from the original graph."""
    euler_circuit = []
    naive_circuit = list(nx.eulerian_circuit(graph_augmented, source=starting_node))

    for edge in naive_circuit:
        edge_data = graph_augmented.get_edge_data(edge[0], edge[1])    

        if edge_data[0]['attr_dict']['trail'] != 'augmented':
            # If `edge` exists in original graph, grab the edge attributes and add to eulerian circuit.
            edge_att = graph_original[edge[0]][edge[1]]
            euler_circuit.append((edge[0], edge[1], edge_att))
        else:
            aug_path = nx.shortest_path(graph_original, edge[0], edge[1], weight='distance')
            aug_path_pairs = list(zip(aug_path[:-1], aug_path[1:]))

            print(u'Fülle Kanten für erweiterte Kanten aus: {}'.format(edge))
            print('Erweitere Pfad: {}'.format(' => '.join(aug_path)))
            print('Erweitere Pfadpaare: {}\n'.format(aug_path_pairs))

            # If `edge` does not exist in original graph, find the shortest path between its nodes and
            #  add the edge attributes for each link in the shortest path.
            for edge_aug in aug_path_pairs:
                edge_aug_att = graph_original[edge_aug[0]][edge_aug[1]]
                euler_circuit.append((edge_aug[0], edge_aug[1], edge_aug_att))

    return euler_circuit
```

```python
def create_cpp_edgelist(euler_circuit):
    """
    Create the edgelist without parallel edge for the visualization
    Combine duplicate edges and keep track of their sequence and # of walks
    Parameters:
        euler_circuit: list[tuple] from create_eulerian_circuit
    """
    cpp_edgelist = {}

    for i, e in enumerate(euler_circuit):
        edge = frozenset([e[0], e[1]])

        if edge in cpp_edgelist:
            cpp_edgelist[edge][2]['sequence'] += ', ' + str(i)
            cpp_edgelist[edge][2]['visits'] += 1

        else:
            cpp_edgelist[edge] = e
            cpp_edgelist[edge][2]['sequence'] = str(i)
            cpp_edgelist[edge][2]['visits'] = 1

    return list(cpp_edgelist.values())
```
