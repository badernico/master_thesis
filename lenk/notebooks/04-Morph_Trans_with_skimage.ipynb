# Test of Morpholgical transformation with skimage

The current morpholgical transformation applied in the tropy optical_flow package is tested again the transformation within skimage.

```python
%matplotlib inline

import pylab as pl
import numpy as np
import datetime
from l15_msevi.msevi import MSevi
import analysis_tools.optical_flow as oflow
```

## Get Example Data

```python
t = datetime.datetime(2013,6,20,11,15)
plon, plat = 9.3, 50.3
scan_type = 'rss'
bsize = 51 # the box size of the cutout

bc = []
for i in range(2):
    
    # read sat-data
    s = MSevi(time = t, scan_type = scan_type, region = 'eu')
    s.lonlat()
    s.load('IR_108')
    s.rad2bt()
    
    # cutout
    p = plon, plat
    ind = gi.get_index(p, s.lon, s.lat)

    clon = gi.cutout_field4box(s.lon, ind, bsize)
    clat = gi.cutout_field4box(s.lat, ind, bsize)
    bc.append( gi.cutout_field4box(s.bt['IR_108'], ind, bsize) )
    
    t += datetime.timedelta(minutes = 30)

bc = np.array( bc)
```

```python
pl.figure(figsize = (16,6))
pl.subplot(121)
pl.pcolormesh(clon, clat, bc[0])

pl.subplot(122)
pl.pcolormesh(clon, clat, bc[1])
```
