```python
def track_reader(fname):
    
    # open netcdf
    f = netCDF4.Dataset(fname, 'r')
    v = f.variables
    
    lon = v['longitude'][:]
    lat = v['latitude'][:]
    time = v['time'][:]
    
    f.close()
    
    # convert time
    t0 = datetime.datetime(1,1,1,0,0)
    
    tobj = []
    
    for t in time:
        tobj.append( t0 + datetime.timedelta(seconds = t))
    
    
    return tobj, lon, lat
```

```python
t, tlon, tlat = track_reader(fname)

track = dict(time = t, lon = tlon, lat = tlat)
```

## Gather Satellite Data

Next, we like to get the satellite data. The steps are
 * open satellite data container and reader BT data for a certain time
 * cutout the region 
 * stack the data in time

```python
def get_satstack(track, bsize = 51):
    
    sat_stack = {}
    for vname in ['lon', 'lat', 'bt108']:
        sat_stack[vname] = []
    
    # time loop
    for i, t in enumerate( track['time'] ):
        
        print '...reading sat.-data for ', t
        
        # setup sat.-container
        s = MSevi(time= t, scan_type='rss')
        
        # read data
        s.lonlat()  # it is quite ineffective to read geo-ref. each time ...
        s.load('IR_108')
        s.rad2bt()
        
        # get index of the track
        p = track['lon'][i], track['lat'][i]
        print p
        
        ind = gi.get_index(p, s.lon, s.lat)
        
        
        # do cutout
        bsize = 51 # the box size of the cutout

        clon = gi.cutout_field4box(s.lon, ind, 51)
        clat = gi.cutout_field4box(s.lat, ind, 51)
        bc = gi.cutout_field4box(s.bt['IR_108'], ind, 51)
        
        
        # gather the field into the dictionary
        sat_stack['lon'].append ( clon )
        sat_stack['lat'].append ( clat )
        sat_stack['bt108'].append ( bc )
        
        
    # finally make arrays
    for vname in ['lon', 'lat', 'bt108']:
        sat_stack[vname] = np.array( sat_stack[vname] )
    
    return sat_stack
```

```python
sdat = get_satstack(track)
```
