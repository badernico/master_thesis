```python
def remap_grid_to_res(grid_data,grid_lon=None,grid_lat=None,
                      new_lon=None,new_lat=None,
                      interpolation_type='nneighbour',
                          index_set=None,**kwargs):
    """
    Remaps a given grid to the RADOLAN grid.
    
    Parameters
    ----------
    grid_data: numpy array, 2d
        grid data to remap
        
    grid_lon: numpy array, default = None
        longitude of the grid cells, only needed if no index_set is given to
        calculate new index set
        
    grid_lat: numpy array, default = None
        latitude of the grid cells, only needed if no index_set is given to
        calculate new index set
        
    interpolation_type:  string, default = 'nneighbour'
        type of the interpolation to use, valid 'nneighbour','curve_flow'
                         
    index_set:  numpy array, 2d, default = None,
        set of indices to remapgrid data to radolan grid, can be calculated if
        grid cell pixel coorinates are given
        
    kwargs: arguments to be passed on to used functions
    
    Returns
    -------
    grid data remapped and the RADOLAN domain
    """
    
    valid_interpolations = ['nneighbour','curve_flow']
        
    if interpolation_type in valid_interpolations:
        # load Radolan coordinates
        #r = rado.Radolan()  
        
        
        # if there is no index set for mapping create one (but it is slow)
        if np.any(index_set)!= None and np.any(grid_lon)!=None and np.any(grid_lat)!=None:
            index_set = gi.create_interpolation_index(grid_lon, grid_lat, 
                                                      new_lon, new_lat)
        
        # interpolate SEVIRI data to RADOLAN grid
        if interpolation_type == 'nneighbour':
            grid_remapped = grid_data[index_set]

            return grid_remapped
        
        elif interpolation_type == 'curve_flow':
            grid_remapped = grid_data[index_set]
            grid_remapped = gi.curve_flow_filter(grid_remapped,3)
            return grid_remapped
        
        else:
            message = "Interpolation error."
            print('\x1b[1;33;41m' + message + '\x1b[0m')
            return
        
    else:
        message = "{} is not a valid interpolation type. Only".format(interpolation_type) + \
                  " 'nneighbour', 'idw' and 'curve_flow' are valid."
        print('\x1b[1;33;41m' + message + '\x1b[0m')
        return
```
