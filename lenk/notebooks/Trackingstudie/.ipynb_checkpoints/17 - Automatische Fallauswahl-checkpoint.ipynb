- eine Mindestlebensdauer von 30 Minuten haben.

```python
def filter_haci(haci_file_name):
    haci_case_data = pd.read_table(haci_file_name,sep=r"\s+")
    
    date = haci_file_name.split("/")[-1].split(".")[0][5:-5]
    
    # Zeitbereich auswählen
    day_time_range = times_range[(times_range.day==int(date[-2:])) & (times_range.month == int(date[5:7]))]
    
    start_time_idx = int(day_time_range.start_hour.values) * 12 + int(day_time_range.start_minute.values) // 5
    end_time_idx = int(day_time_range.end_hour.values) * 12 + int(day_time_range.end_minute.values) // 5
    
    haci_case_data_enough_light = haci_case_data[(haci_case_data.t0 >= start_time_idx) & \
                                                 (haci_case_data.t1 <= end_time_idx)]
    
    haci_case_long_enough = haci_case_data_enough_light[haci_case_data_enough_light.t1 - \
                                                        haci_case_data_enough_light.t0 >=5]
    return haci_case_long_enough
```

```python
def convert_haci_bbox2msg(object_data,box_size=35):
    from io_tools import radolan
    from analysis_tools import grid_and_interpolation as gi
    
    r = radolan.Radolan()
    slon,slat = mst.get_msg_lon_lat('eu')
    
    centroid_line = object_data.l00 + (object_data.l01 - object_data.l00) / 2.
    centroid_column = object_data.c00 + (object_data.c01 - object_data.c00) / 2.
    
    centroid_lon = r.lon[int(np.rint(centroid_line)),int(np.rint(centroid_column))]
    centroid_lat = r.lat[int(np.rint(centroid_line)),int(np.rint(centroid_column))]
    
#     box_lon_min = r.lon[int(np.rint(object_data.l0)),
#                         int(np.rint(object_data.c0))]
    
#     box_lat_min = r.lat[int(np.rint(object_data.l0)),
#                         int(np.rint(object_data.c0))]
    
#     box_lon_max = r.lon[int(np.rint(object_data.l1)),
#                         int(np.rint(object_data.c1))]
    
#     box_lat_max = r.lat[int(np.rint(object_data.l1)),
#                         int(np.rint(object_data.c1))]


    
    sat_row, sat_col = gi.get_index((centroid_lon,centroid_lat),slon,slat)
    sat_row_min = np.clip(sat_row - (box_size // 2),0,900)
    sat_row_max = np.clip(sat_row + (box_size // 2) + 1,0,900)
    sat_col_min = np.clip(sat_col - (box_size // 2),0,900)
    sat_col_max = np.clip(sat_col + (box_size // 2) + 1,0,900)
    
    sat_lon_min = slon[sat_row_min,sat_col_min]
    sat_lat_min = slat[sat_row_max,sat_col_min]
    sat_lon_max = slon[sat_row_max,sat_col_max]
    sat_lat_max = slat[sat_row_min,sat_col_max]
    
    out_dict = {'lon_min':sat_lon_min,
                'lat_min':sat_lat_min,
                'lon_max':sat_lon_max,
                'lat_max':sat_lat_max,
                'row_min':sat_row_min,
                'row_max':sat_row_max,
                'col_min':sat_col_min,
                'col_max':sat_col_max}
    
    return out_dict    
```

```python
def plot_haci_cases(case):
    t_start = "{}T{:02d}{:02d}".format(case.date.replace("-",""),(case.t0*5)//60, (case.t0*5) % 60)
    t_end = "{}T{:02d}{:02d}".format(case.date.replace("-",""),(case.t1*5)//60, (case.t1*5) % 60)
    times = pd.date_range(start=t_start,end=t_end,freq="5min")

    cutouts = []

    for t in times:
        r = radolan.Radolan()
        r.read(t.to_pydatetime())
        
        cutouts.append(r.data[np.clip(case.l0-10,0,900):np.clip(case.l1+10,0,900),
                              np.clip(case.c0-10,0,900):np.clip(case.c1+10,0,900)])
        
    fig,ax = plt.subplots(1,len(times),figsize=(len(times)*3,6),sharey=True)

    for i,t in enumerate(times):
        ax[i].imshow(np.ma.masked_less(np.ma.masked_greater_equal(cutouts[i],92.5),5),cmap=rc(),vmin=5,vmax=70)
        ax[i].set_title(t.strftime("%d.%m.%Y %H:%M UTC"))

    plt.tight_layout()
    plt.savefig("{}/{}_{}_{}.png".format(pic_path, case.date,case.id, case.time.replace(":","")))
    plt.close("all")
    return
```

```python
def duration_filter(objprops, min_dur = 6):

    '''
    Filter for persisent Objects.
    '''


    t_duration = objprops.t1 - objprops.t0
    objprops['t_duration'] = t_duration

    return t_duration >= min_dur

def range_filter(objprops, field_shape, 
                 time_edge = 12, 
                 space_edge = 30):

    '''
    Filters objects too close to bounds.
    '''


    ntime, nrow, ncol = field_shape

    # temporal filter
    m_time1 = (objprops.t0 > time_edge)
    m_time2 = (objprops.t1 < (ntime - time_edge))
    
    m_time_range = m_time1 & m_time2


    # Filtering based on Bounding Box
    
    # lines or rows
    m_lrange = (objprops.l00 > space_edge) &  (
        objprops.l01 < nrow -  space_edge)
    
    # columns
    m_crange = (objprops.c00 > space_edge) &  (
        objprops.c01 < ncol -  space_edge)

    
    # combination of all
    m_range = m_lrange & m_crange & m_time_range

    return m_range


def area_growth_filter(objprops,
                       max_init_area = 100,
                       min_area_change = 200):

    '''
    Calculates and compares areas from bounding boxes.
    '''


    area0 = (objprops.l01 - objprops.l00) * (objprops.c01 - objprops.c00)
    area_max = (objprops.l1 - objprops.l0) * (objprops.c1 - objprops.c0)

    objprops['area0'] = area0
    objprops['area_max'] = area_max

    # area change
    darea = area_max - area0
    objprops['darea'] = darea


    m_area = (area0 < max_init_area) & (darea > min_area_change)

    return m_area

def filter_20dbz(objprops,min_area_ratio):
    """
    Calculates and compares the ratio of the area with more than 35 dBZ and more than 20 dBZ
    """
    from skimage import measure
    
    objprops['area_ratio'] = 0
    r = radolan.Radolan()
    
    rr = pd.DataFrame(index = objprops.index,columns=['area_ratio'])
    for i, obj in objprops.iterrows():
        date_t0 = "{}T{:02d}{:02d}".format(obj.date.replace("-",""),(obj.t0*5)//60, (obj.t0*5)%60)
        date_t0 = pd.Timestamp(date_t0)
        
        r.read(date_t0.to_pydatetime())
        
        cutout = r.data[obj.l0:obj.l1,obj.c0:obj.c1]
        
        pixel_20dbz = np.where(cutout>=20)
        pixel_35dbz = np.where(cutout>35)
        
        ratio = float(len(pixel_35dbz[0])) / len(pixel_20dbz[0])
        
        rr.loc[i].area_ratio = ratio
        
    m_ratio = rr.area_ratio > min_area_ratio
    
    return rr,m_ratio

def lightning_filter(objprops):
    """
    Loads and compares the lighning density along the object track.
    """
    
    lightning_present = pd.DataFrame(index = objprops.index,columns=['lightning'])
    
    for i, obj in objprops.iterrows():
        date_t0 = "{}T{:02d}{:02d}".format(obj.date.replace("-",""),((obj.t0-5)*5)//60, (obj.t0*5)%60)
        date_t1 = "{}T{:02d}{:02d}".format(obj.date.replace("-",""),(obj.t1*5)//60, (obj.t1*5)%60)
        
        time_range = pd.date_range(start = date_t0, end=date_t1,freq="5min")
        
        for i,t in enumerate(time_range):
            if i < len(time_range)-1:
                linet_data = rld.read_linet_for_time_intervall(t.to_pydatetime(),time_range[i+1].to_pydatetime())
                ld = rld.linet2rado_frequency_grid(linet_data)
                
                if np.any(np.array(ld[obj.l0:obj.l1,obj.c0:obj.c1])!=0):
                    lightning_present.iloc[i].lightning = 1
            else:
                continue
    
    return lighning_present
```
