```python
import numpy as np
import pandas as pd
import glob

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
stat_folder ="/vols/satellit/home/lenk/proj/2019-01_trackingstudie/data/cutouts_t0/stats/"
```

```python
stat_files = glob.glob("{}*.csv".format(stat_folder))
```

```python
metrics = {'mean':np.nanmean,'median':np.nanmedian,'min':np.nanmin,
           'max':np.max,'q25':np.nanpercentile,'q75':np.nanpercentile}
```

```python
stat_collection = {f:{k:[] for k in metrics.keys()} for f in ['IR_108','CTTH_HEIGHT']}
```

```python
field_type = []
stat_type = []
values = []

for sf in stat_files:
    stat_data = pd.read_csv(sf,index_col=0)
    
    for field in ['IR_108','CTTH_HEIGHT']:
        for imetric in metrics.keys():
            field_type.append(field)
            stat_type.append(imetric)
            values.append(stat_data[field][imetric])
```

```python
stat_collection = pd.DataFrame({'stat_type':stat_type,
                                'field_type':field_type,
                                'values':values})
```

```python
stat_collection
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,8))
sns.boxplot('stat_type','values',data=stat_collection[stat_collection.field_type=='IR_108'],ax=ax)
ax.set_title(r"Cloud object properties at $t_\mathrm{CI}$, N=329")
ax.set_xlabel("")
ax.set_ylabel(r"$T_\mathrm{B}$" + u"(10.8 Âµm)")
plt.savefig("/vols/satellit/home/lenk/proj/2019-01_trackingstudie/pics/cutout_t0_stat/IR_108.pdf")
```
