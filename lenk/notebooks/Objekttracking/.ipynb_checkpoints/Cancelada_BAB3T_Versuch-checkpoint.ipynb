# Objekttracking nach BAB3T (Cancelada et al., 2020)

In dem Aufsatz von Cancelada et al. (2020), gibt es einen Interessanten Ansatz für die Rückverfolgung von konvektiven Systemen mit adatpiven Schwellwerten. Die Grundprinzipien sind folgende:

1. Es werden Wolkencluster mit einem Satz von Schwellwerten konstruiert, die von 215&thinsp;K bis 235&thinsp;K reichen.
2. Zu den Clustern werden Niederschlagsobjekte in einem Zeitraum von &pm;&thinsp;30&thinsp;min und einer Entfernung von maximal 16&thinsp;km ausgewählt. Es wird die nächstkältere mögliche Gruppierung ausgewählt, wenn in der Überschneidung von Niederschlagsschwerpunkt und Wolkencluster noch ein kälteres Cluster enthalten ist.
3. Als nächstes wird im vorherigen Zeitschritt nach einem überlappenden Cluster mit gleicher Temperatur gesucht. Wenn Systeme verschmolzen sind, wird das mit dem größten Überlapp ausgewählt.
4. Wenn keine Überschneidung mit derselben Temperatur gefunden wird, wird nach einer mit höherer Tempratur gesucht.
5. Das wird solange durchgeführt, bis für ein Cluster mit 235&thinsp;K im vorherigen Zeitschritt keine Überklappung mehr gefunden wird.
6. Um Sprünge wegen Verschmelzungen zu vermeiden, wird eine maximale Entfernung von 200&thinsp;km zwischen den Clusterschwerpunkten zweier aufeinander folgender Bilder festgelegt. Wenn die Entfernung größer ist, wird im nächsten vorherigen Zeitschritt gesucht. Wenn auch da nichts gefunden wird, wird die Rückverfolgung abgebrochen.

Das probieren wir doch mal aus.

```python
import numpy as np
import datetime as dt
import pandas as pd
import load_satellite_data as lsd
import matplotlib.pyplot as plt
%matplotlib inline
from plotting_tools.colormaps import enhanced_colormap
emap = enhanced_colormap(vmin=210,vmed=240,vmax=300)
import fixed_colourbar as fc
from scipy import ndimage as ndi
```

```python
t0 = dt.datetime(2013,6,18,13,45)
t1 = dt.datetime(2013,6,18,13,50)
t2 = dt.datetime(2013,6,18,13,55)
t3 = dt.datetime(2013,6,18,14,0)
```

```python
tlist = [t3,t2,t1,t0]
```

```python
thresholds = [215,218,223,235,245]
```

```python
ir108 = np.zeros((4,165,300))
```

```python
for i,t in enumerate(tlist):
    ir108[i,:,:] = lsd.load_satellite_data(t,'IR_108')[90:255,340:640]
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,8))
p = ax.imshow(ir108[0],vmin=210,vmax=300,cmap=emap)
fc.colourbar(p)
ax.set_title("IR 10,8 µm, {}".format(tlist[0]))
```

```python
masked = np.zeros((4,165,300))
objects = np.full((4,165,300),False)
```

```python
for i,ir in enumerate(ir108):
    M = np.full_like(ir,0)
    for th in thresholds[::-1]:
        m  = np.ma.masked_less_equal(ir,th)
    
        M[np.where(m.mask==True)] = th
    
    masked[i,:,:] = M
    objects[i,:,:] = ~np.ma.masked_less(M,1).mask
```
