# Erweiterung der Kostenfunktion nach Lakshmanan & Smith (2010)

Die in der Veröffentlichung von Lakshmanan & Smith (2010) beschriebene Kostenfunktion zum Objekttracking ist ein guter Startpunkt, aber ihr fehlen ein paar Punkte.

Sie ist folgendermaßen definiert:
$$ c_{i,j}  = \left ( x_i - x_j\right ) ^2  + \left ( y_i - y_j\right ) ^2  + \frac{A_j}{\pi} \left ( \frac{\left| A_i - A_j\right | } {\mathrm{max}\left( A_i, A_j\right)} + \frac{\left| d_i - d_j\right | } {\mathrm{max}\left( d_i, d_j\right)} \right ) $$

Im dritten Term geht die Fläche zweimal ein, einmal über die Fläche selbst und nochmal über den Durchmesser, aber das ist wahrscheinlich gut so.

In der Gleichung ist allerdings keine Überschneidung zwischen Objekten enthalten. Die sollten wir noch hinzufügen.

## 1. Objektüberschneidung

Wenn es mehrere Kandidaten gibt, die sich überschneiden, sollten wir diese Überschneidung mit in die Formel aufnehmen. Die Überschneidung ist definiert als: $ A \cap B := \left{ x | (x \in A) \wedge (x \in B) \right} $. Dieser Wert sollte maximal sein. Da die Kostenfunktion allerdings darauf ausgelegt ist, minimiert zu werden, müssen wir eine andere Relation finden. Es bietet sich dafür an, den Bereich zu betrachten, in dem sich die Objekte nicht überschneiden. Wenn die Überschneidung größer wird, wird dieser Bereich kleiner und sollte also minimiert werden. Dieser Bereich wird als symmetrische Differenz bezeichnet und ist definiert als: $ A \triangle B := \left{ x | (x \in A) \veebar (x \in B) \right} = A \cup B - A \cap B $. Anschließend kann der Wert noch ähnlich, wie im Term mit der Fläche und dem Durchmesser mit dem Maximum des jeweiligen überschneidungsfreien Bereich normiert werden:

$$ d = \frac{A_i \triangle A_j}{\mathrm{max}\left [ \left ( A_i \setminus A_j \right ), \left (A_j \setminus A_i \right )\right ] } $$

Die Kostenfunktion sähe dann also wie folgt aus:

$$ c_{i,j}  = \left (x_i - x_j\right ) ^2  + \left ( y_i - y_j\right ) ^2  + \frac{A_j}{\pi} \left ( \frac{\left| A_i - A_j\right | } {\mathrm{max}\left( A_i, A_j\right)} + \frac{\left| d_i - d_j\right | } {\mathrm{max}\left( d_i, d_j\right)} \right )  + \frac{A_i \triangle A_j}{\mathrm{max}\left [ \left ( A_i \setminus A_j \right ), \left (A_j \setminus A_i \right )\right ] } $$

Wenn zwei Objekte identisch sind, wird $\mathrm{max}\left [ \left ( A_i \setminus A_j \right ), \left (A_j \setminus A_i \right )\right ]$ null und damit ist der Überschneidungsterm undefiniert. Um auch mit solchen Fällen umgehen zu können, ist eine weitere Anpassung der Kostenfunktion nötig. Eine Variante könnte sein den Überschneidungsterm aus dem Verhältnis von symmetrischen Differenz und Überschneidungsfläche aufzubauen. Dieses Verhältnis sollte ebenfalls gegen null gehen, wenn die Überschneidung größer wird:

$$ c_{i,j}  = \underbrace{\left ( x_i - x_j\right ) ^2 + \left ( y_i - y_j\right ) ^2}_\text{Entfernung der Schwerpunkte}  + \underbrace{\frac{A_j}{\pi} \left ( \frac{\left| A_i - A_j\right | } {\mathrm{max}\left( A_i, A_j\right)} + \frac{\left| d_i - d_j\right | } {\mathrm{max}\left( d_i, d_j\right)} \right )}_\text{Flächenunterschied}  + \underbrace{\frac{A_i \triangle A_j}{A_i \land A_j  }}_\text{Überschneidung} $$

Wenn es keine Überschneidung gibt, wird ${A_i \land A_j}$ null und damit ist der Term nicht definiert. Da wir nur Objekte mit einer gemeinsamen Überschneidung verfolgen wollen, ist das in Ordnung. Die Kostenfunktion ist dann nicht definiert und wir bekommen für sich nicht überschneidende Objekte keinen Wert. 

## 2.  Test der Kostenfunktion

Für einen Test sehen wir uns mal ein einfaches Beispiel an. Wir haben zwei elliptische Objekte, die sich überschneiden:

* Objekt 1 mit folgenden Koordinaten: $x = 14$, $y = 15$, $d_x = 12\,\mathrm{px}$, $d_y = 10\,\mathrm{px}$
* Objekt 2 mit folgenden Koordinaten: $x = 20$, $y = 28$, $d_x = 10\,\mathrm{px}$, $d_y = 12\,\mathrm{px}$

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import ellipse
```

```python
def equiv_diameter(A):
    return np.sqrt((4*A) / np.pi)

def intersections(A,B):
    A_area = np.count_nonzero(A)
    B_area = np.count_nonzero(B)

    intersection_area = np.count_nonzero(A & B)
    symmetrical_difference = np.count_nonzero(A ^ B)

    #A_not_B_area = np.count_nonzero(A ^ (A & B))
    #B_not_a_area = np.count_nonzero(B ^ (A & B))
    
    return (intersection_area, symmetrical_difference, A_area, B_area)#,A_not_B_area, B_not_a_area)

def intersection_term(object0,object1):
    """
    Derive intersection term for the object tracking cost function.
    
    Inputs:
        * object0: array-like, int, range = [0,1]
            object to look for fitting sucessor candidates
        * object1: array-like, int, range = [0,1]
            possible successir candidate for object0
            
    Returns:
        * intersection term
    """
    # derive intersection, symmetrical difference, object areas
    (intersection,difference,object0_area,object1_area) = intersections(object0,object1)
    
    try:
        # calculate and return intersection term
        #return ((object0_area + object1_area) / np.pi) * (difference / intersection)
        return difference / intersection
    except:
        # if there is no instersection the term in not defined, thus return np.nan
        return np.nan

def euclidian_distance(point0,point1):
    """
    Calculate the Euclidian distance of two points
    """
    
    distance = 0
    
    for i in range(len(point0)):
        distance = distance + (point0[i] - point1[i])**2
        
    return np.sqrt(distance)

def cost(x0,x1,y0,y1,A0,A1,obj0,obj1):
    dx = (x0-x1)**2
    dy = (y0-y1)**2

    dA = np.abs(A0 - A1) / np.max((A0,A1))

    d0 = equiv_diameter(A0)
    d1 = equiv_diameter(A1)
    dd = np.abs(d0 - d1) / np.max((d0,d1))

    dA = (A1 / np.pi) * (dA + dd)

    dU = intersection_term(obj0,obj1)

    return dx + dy + dA + dU
```

```python
objekt1 = {'x':14,'y':15,'dx':12,'dy':10}
objekt2 = {'x':20,'y':28,'dx':10,'dy':12}
```

```python
field1 = np.full((40,40),False)
field2 = np.full((40,40),False)

rr1,cc1 = ellipse(objekt1['x'],objekt1['y'],objekt1['dx'],objekt1['dy'])
rr2,cc2 = ellipse(objekt2['x'],objekt2['y'],objekt2['dx'],objekt2['dy'])

field1[rr1,cc1] = True
field2[rr2,cc2] = True
```

```python
fig,ax = plt.subplots(3,2,figsize=(10,15))
axs = ax.ravel()
axs[0].imshow(field1)
axs[0].set_title("Objekt 1")
axs[0].plot(15,14,'ro',mew=1.5,markersize=8)
axs[1].imshow(field2)
axs[1].set_title("Objekt 2")
axs[1].plot(28,20,'ro',mew=1.5,markersize=8)
axs[2].imshow(field1 & field2)
axs[2].set_title("Überschneidung")
axs[3].imshow(field1 ^ field2)
axs[3].set_title("symmetrische Differenz")
axs[4].imshow(field1 ^ (field1 & field2))
axs[4].set_title("Objekt 1 \ Objekt 2")
axs[5].imshow(field2 ^ (field1 & field2))
axs[5].set_title("Objekt 2 \ Objekt 1")
```

```python
A_objekt1 = np.count_nonzero(field1)
A_objekt2 = np.count_nonzero(field2)

A_schnitt = np.count_nonzero(field1 & field2)
A_diff = np.count_nonzero(field1 ^ field2)

A_obj1_nicht_obj2 = np.count_nonzero(field1 ^ (field1 & field2))
A_obj2_nicht_obj1 = np.count_nonzero(field2 ^ (field1 & field2))
```

Objekt 1 und Objekt 2 haben jeweils eine Fläche von {{"{} px²".format(A_objekt1)}}, also zusammen {{"{} px²".format(A_objekt1 + A_objekt2)}}. Die Überschneidungsfläche beträgt {{"{} px²".format(A_schnitt)}} und die Fläche der symmetrischen Differenz {{"{} px²".format(A_diff)}}. Der Überschneidungsterm der Kostenfunktion sieht also wie folgt aus:

$$ d = \frac{\mathrm{Objekt 1} \triangle \mathrm{Objekt 2}}{ \mathrm{Objekt 1} \wedge \mathrm{Objekt 2}}   = \frac{570\,\mathrm{px}^2}{84\,\mathrm{px}^2} \approx 6,8$$

$$ \frac{A_i + A_j}{\pi} \cdot d = \frac{738\,\mathrm{px}}{\pi} \cdot 2 \approx 470 $$

```python
c = cost(objekt1['x'],objekt2['x'],objekt1['y'],objekt2['y'],A_objekt1,A_objekt2,field1,field2)
```

In diesm Fall würde Objekt 3 als Fortsetzung von Objekt 2 angenommen, weil die Kostenfunktion dafür kleiner ist. Das sollen wir im Detail untersuchen, um zu verstehen warum.

```python
dx_o1_o2 = np.abs(objekt1['x'] - objekt2['x'])**2
dy_o1_o2 = np.abs(objekt1['y'] - objekt2['y'])**2

#######################################################
A_objekt1 = np.count_nonzero(field1)
A_objekt2 = np.count_nonzero(field2)
dA_o1_o2 = np.abs(A_objekt1 - A_objekt2) / np.max((A_objekt1, A_objekt2))

#######################################################
d_o1 = equiv_diameter(A_objekt1)
d_o2 = equiv_diameter(A_objekt2)
dd_o1_o2 = np.abs(d_o1 - d_o2) / np.max((d_o1,d_o2))

dA_ges_o1_o2 = (A_objekt1 / np.pi) * (dA_o1_o2 + dd_o1_o2)

#######################################################
A_schnitt_o1_o2 = np.count_nonzero(field1 & field2)
A_diff_o1_o2 = np.count_nonzero(field1 ^ field2)

#A_obj1_nicht_obj2 = np.count_nonzero(field1 ^ (field1 & field2))
#A_obj2_nicht_obj1 = np.count_nonzero(field2 ^ (field1 & field2))
      
dU_o1_o2 = (A_diff_o1_o2 / A_schnitt_o1_o2) # *((A_objekt1 + A_objekt2) / np.pi)
```

Für Objekt 1 und Objekt 2 sehen die Werte der Kostenfunktionsterme wie folgt aus:

- $\frac{\left | d_\mathrm{Objekt 1} - d_\mathrm{Objekt 2}\right |}{\mathrm{max} \left( d_\mathrm{Objekt 1}, d_\mathrm{Objekt 2} \right ) }$ = {{"{:3.1f} px".format(dd_o1_o2)}}, weil die Objekte gleich groß sind

Die Kostenfunktion ergibt für die möglichen Kombinationen folgende Werte:
- c(Objekt 1, t0 &rarr; Objekt 1, t1) = {{"{:3.2f}".format(c1_1)}}
- c(Objekt 1, t0 &rarr; Objekt 2, t1) = {{"{:3.2f}".format(c1_2)}}
- c(Objekt 2, t0 &rarr; Objekt 2, t1) = {{"{:3.2f}".format(c2_2)}}

Das heißt, dass Objekt 1 im zweiten Zeitschritt als Vortsetzung von Objekt 1 im ersten Zeitschritt und Objekt 2 im zweiten Zeitschritt als Fortsetzung von Objekt 2 im ersten Zeitschritt genommen würde. Objekt 1  im ersten Zeitschritt und Objekt 2 im zweiten Zeitschritt haben keine Übeschneidung und deshalb ist die Kostenfunktion null. Das passt.

```python
def find_successor_lakshmanan(objects_t0,objects_t1,t0,t1,search_factor=1):
    # find object properties
    centroids_t0, areas_t0, diameters_t0 = get_object_properties4tracking(objects_t0)
    centroids_t1, areas_t1, diameters_t1 = get_object_properties4tracking(objects_t1)

    # 
    
    # connect objects
    connections = dict()
    for i0, c0 in enumerate(centroids_t0):
        # extract object centroid, area, diameter and label
        x0 = c0[0]
        y0 = c0[1]
        area0 = areas_t0[i0]
        diameter0 = diameters_t0[i0]
        
        # determine object label
        label0 = i0 + 1
        
        cost = dict()
        for i1, c1 in enumerate(centroids_t1):
            x1 = c1[0]
            y1 = c1[1]
            area1 = areas_t1[i1]
            diameter1 = diameters_t1[i1]
            
            # calculate centroid distance
            dist = np.linalg.norm(np.array([x0,y0])-np.array([x1,y1]))
            
            # check if centroid is within the search radius
            if dist <= diameter0 * search_factor:
                # determine object label
                label1 = i1 + 1
                
                # calculate cost function
                cost["{:03d}_{:04d}".format(t1,label1)] = centroid_cost(x0,x1,y0,y1,area0,area1,diameter0,diameter1)
     
        # look for the object with the least cost, that is the successor
        try:
            min_idx = np.argmin(np.array(list(cost.values())))

            connections["{:03d}_{:04d}".format(t0,label0)] = [list(cost.keys())[min_idx]]

        except:
            continue
            
    return connections
```
