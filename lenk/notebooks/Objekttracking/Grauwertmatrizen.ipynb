# Grauwertmatrizen

Grauwertmatrizen basieren auf der Texturinformation und sind eine Möglichkeit ein Grauwertbild zu segmentieren. Vielleicht eignen sie sich auch, um MSG-Bilder zu segmentieren. Das probieren wir hier mal aus.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.feature import greycomatrix, greycoprops
from skimage import data

import sys
sys.path.append("/vols/talos/home/stephan/utils/tracking")
import tracking_common as tc

import l15_msevi.msevi as msv
import datetime as dt
```

Zuerst sehen wir uns mal das Beispiel von der Scikit-Image-Internetseite an. Dafür definieren wir zunächst die Größe der Bereiche, für die wir die Texturinformationen analysieren. In diesem Fall benutzen wir Ausschnitte von 21 x 21 px.

```python
PATCH_SIZE = 21
```

Dann laden wir das Beispielbild mit dem Kameramann.

```python
image = data.camera()

fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.imshow(image,vmin=0,vmax=255,cmap='gray')
```

Dann wählen wir einige Bereiche auf der Wiese im Bild aus.

```python
grass_locations = [(474, 291), (440, 433), (466, 18), (462, 236)]
grass_patches = []
for loc in grass_locations:
    grass_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])
```

Und dann noch einige Bereiche aus dem Himmel im Bild.

```python
sky_locations = [(54, 48), (21, 233), (90, 380), (195, 330)]
sky_patches = []
for loc in sky_locations:
    sky_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                             loc[1]:loc[1] + PATCH_SIZE])
```

Dann sehen wir uns mal das Bild und die Ausschnittsbereiche an.

```python
fig,ax = plt.subplots(1,1,figsize=(8, 8))
ax.imshow(image, cmap=plt.cm.gray, interpolation='nearest',
          vmin=0, vmax=255)
for (y, x) in grass_locations:
    ax.plot(x + PATCH_SIZE / 2, y + PATCH_SIZE / 2, 'gs')
for (y, x) in sky_locations:
    ax.plot(x + PATCH_SIZE / 2, y + PATCH_SIZE / 2, 'bs')
ax.set_xlabel('Originalbild')
ax.set_xticks([])
ax.set_yticks([])
ax.axis('image')
```

Und die Ausschnitte einzeln.

```python
fig = plt.figure(figsize=(20,10))
# display the image patches
for i, patch in enumerate(grass_patches):
    ax = fig.add_subplot(3, len(grass_patches), len(grass_patches)*1 + i + 1)
    ax.imshow(patch, cmap=plt.cm.gray, interpolation='nearest',
              vmin=0, vmax=255)
    ax.set_xlabel('Gras %d' % (i + 1))

for i, patch in enumerate(sky_patches):
    ax = fig.add_subplot(3, len(sky_patches), len(sky_patches)*2 + i + 1)
    ax.imshow(patch, cmap=plt.cm.gray, interpolation='nearest',
              vmin=0, vmax=255)
    ax.set_xlabel('Himmel %d' % (i + 1))
```

Wir zu sehen ist, unterscheiden sich die Gras- und Himmelsauschnitte deutlich in ihrer Textur. Die Grasflächen sind inhomogener während die Himmelsausschnitte relativ monotone Grauwerte aufweisen. Als nächstes berechnen wir einige Eigenschaften der Grauwertmatrizen der Ausschnitte.

```python
xs = []
ys = []
for patch in (grass_patches + sky_patches):
    glcm = greycomatrix(patch, [5], [0], 256, symmetric=True, normed=True)
    xs.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    ys.append(greycoprops(glcm, 'correlation')[0, 0])
```

Dann sehen wir uns diese Eigenschaften mal an.

```python
fig,ax = plt.subplots(1,1,figsize=(12,8))
ax.plot(xs[:len(grass_patches)], ys[:len(grass_patches)], 'go',
        label='Gras')
ax.plot(xs[len(grass_patches):], ys[len(grass_patches):], 'bo',
        label='Himmel')
ax.set_xlabel('GLCM-Ungleichheit')
ax.set_ylabel('GLCM-Korrelation')
ax.legend()
```

Für die Himmelsbereiche ist die Korrelation der Grauwerte deutlich höher als für die Grasbereiche, während die Ungleichheit der Grauwerte der Grasbereiche erwartungsgemäß deutlich höher ist. Damit kann man beide Bereiche gut trennen.

```python
glcm = greycomatrix(grass_patches[1], [5], [0], 256, symmetric=True, normed=True)
# greycomatrix(Bild, Entfernungen, Winkel, Schichten=None, symmetrisch=False, genormt=False)

props = {p:{} for p in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']}

for p in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']:
    props[p] = greycoprops(glcm,p)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.imshow(glcm[:, :, 0, 0])
```

```python
land_locations = [(500,1700),(1500,600),(1400,400)]
ocean_locations = [(1000,1400)]
cloud_locations = [(1000,750),(450,2100)]

land_patches = []
for loc in land_locations:
    land_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                              loc[1]:loc[1] + PATCH_SIZE])
    
ocean_patches = []
for loc in ocean_locations:
    ocean_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])
    
cloud_patches = []
for loc in cloud_locations:
    cloud_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])
```

```python
xs = []
ys = []
for patch in (land_patches + ocean_patches + cloud_patches):
    glcm = greycomatrix(patch, [5], [0], 256, symmetric=True, normed=True)
    xs.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    ys.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
fig,ax = plt.subplots(1,len(cloud_patches),figsize=(10,5))

for i,cp in enumerate(cloud_patches):
    ax[i].imshow(cp,vmin=0,vmax=255,cmap='gray')
```
