# Üperprüfung des Validierungsansatzes für die Dissertation

Für das dritte Ergebniskapitel der Disseration wäre es gut den Validierungsansatz nochmals zu überprüfen und eventuell anzupassen.

Der Validierungsansatz besteht aus folgenden Schritten:

1. Wolkenobjekte definieren
2. HACI-Objekte suchen
3. Überprüfen, ob Wolkenobjekte mit CI-Detektionen überlappen
4. Überprüfen ob und wann sich nach einer CI-Detektion ein Niederschlagsobjekt entwickelt

Bei diesem Ansatz können einige Probleme auftreten:

1. Das Hauptproblem ist, dass das das CI-Produkt pixelbasiert und der Validierungsansatz objektbasiert ist. Hier können Inkonsistenzen entstehen.
2. In einem Wolkenobjekt können Pixel mit verschiedenen CI-Wahrscheinlichkeiten liegen.

Als erstes sehen wir uns die Objektdedinition an.

## 1. Objektdefinition

Die Objektdefinition basiert dem HRV-Kanal mit einem festen Schwellwert von 0,3. D.h. alles was eine Reflektivtät von mehr als 0,3 hat, wird als Wolkenobjekt gewertet. Das ist inkonsistent mit dem restlichen Ansatz der Dissertation und wir sollten vergleichen, ob es Unterschiede gibt, wenn wir den sonst verwendeten Objektdefinitionsansatz verwenden.

```python
import numpy as np
import xarray as xr
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from standard_config import *

import sys
sys.path.append("{}/utils/tracking".format(local_home_path))
import segmentation as oseg
import object_tracking as ot
import tracking_common as tco

sys.path.append("{}/2018-05_cieval/python/filtered_graph_branch/".format(proj_path))
import cieval_data_io as cdio
import cieval_tools as ct
import cieval_plotting_tools as cpt

import tqdm
```

Dafür laden wir uns zunächst die Daten für einen der Falltage.

```python
hrv = cdio.load_cieval_data('20130618','hrv','radolan',False)
hrv_objects = cdio.load_cieval_data('20130618','hrv_objects','radolan',False)
ir108 = cdio.load_cieval_data('20130618','ir108','radolan',False)
hrv_flow = cdio.load_cieval_product_flow('20130618','hrv')
ir_flow = cdio.load_cieval_product_flow('20130618','ir108')
```

Als nächstes erzeugen wir uns aus den IR-10,8-µm-Daten Objekte.

```python
ir_objects = []

for ir in tqdm.tqdm(ir108.ir108.data):
    obj = oseg.watershed_local_min_segmentation(ir,7,220,273.15,240,5,0)
    
    ir_objects.append(obj)
```

```python
fig,ax = plt.subplots(3,1,figsize=(8,24))
ax[0].pcolormesh(hrv.lon.data,
                 hrv.lat.data,
                 np.ma.masked_greater(hrv.hrv[48].data,1),vmin=0,vmax=1,cmap='gray')
ax[0].pcolormesh(hrv.lon.data,
                 hrv.lat.data,
                 np.ma.masked_less(hrv_objects.hrv_objects[48].data,1),cmap='nipy_spectral',alpha=0.8)
ax[0].set_title("HRV-Objekte")
ax[1].pcolormesh(ir108.lon.data,
                 ir108.lat.data,
                 ir108.ir108[48].data,vmin=210,vmax=300,cmap='gray_r')
ax[1].pcolormesh(ir108.lon.data,
                 ir108.lat.data,
                 np.ma.masked_less(ir_objects[48],1),cmap='nipy_spectral',alpha=0.8)
ax[1].set_title(u"IR-10,8-µm-Objekte")
ax[2].pcolormesh(ir108.lon.data,
                 ir108.lat.data,
                 np.ma.masked_greater(hrv_objects.hrv_objects[48].data,1).mask*1 +
                 np.ma.masked_greater(ir_objects[48],1).mask*1)
```

Es bestehen einige Unterschiede zwischen den Objektmasken. Wie zu erwarten ist die Auflösung der IR-10,8-µm-Objektmaske größer und es werden auch dünnerer Wolken als Objekte gefunden. Allerdings werden teilweise Objekte, die in der HRV-Maske zu einem Objekt zusammengefasst werden getrennt.

```python
hrv_flow
```

```python
# link the object
hrv_object_connections = {}

for i in range(1,len(hrv_objects.hrv_objects.data)):
    o0 = tco.morph_field(hrv_objects.hrv_objects.data[i-1], 
                         hrv_flow.u[i-1], hrv_flow.v[i-1], method='forward')

    hrv_object_connections[i-1] = ot.find_successor_lakshmanan(o0.astype("int"),
                                                               hrv_objects.hrv_objects.data[i].astype("int"),
                                                               i-1,i,3)

hrv_tracks = ot.link_objects(hrv_object_connections)
```

```python
hrv_tracks
```

```python
# link the object
ir_object_connections = {}

for i in range(1,len(ir_objects)):
    o0 = tco.morph_field(ir_objects[i-1], 
                         ir_flow.u[i-1], ir_flow.v[i-1], method='forward')

    ir_object_connections[i-1] = ot.find_successor_lakshmanan(o0.astype("int"),
                                                              ir_objects[i].astype("int"),
                                                              i-1,i,3)

ir_tracks = ot.link_objects(ir_object_connections)
```

```python
print("HRV-Objekttracks: {}\nIR-Objekttracks:{}".format(len(hrv_tracks),len(ir_tracks)))
```

Für den Beispieltag erhalten wir fast fünfmal soviele IR-Objekte wie HRV-Objekte. Das kann daran liegen, dass der Zeitraum, der vom HRV-Kanal abgedeckt wird kürzer ist. Außerdem werden im Ansatz für den IR-10,8-µm-Kanal auch dünne Wolken als Objekte erkannt, was beim HRV-Kanal durch den Schwellwert eher nicht der Fall ist. Es kann aber auch daran liegen, dass mit dem adaptiven Schwellwert einfach mehr Objekte in sonst größeren Objekten erkannt werden.

```python
ir_tracks
```

```python
hrv_track_lengths = [len(tr) for tr in hrv_tracks]
ir_track_lengths = [len(tr) for tr in ir_tracks]
```

```python
fig,ax = plt.subplots(2,1,figsize=(16,10))
ax[0].hist(hrv_track_lengths,bins=20)
ax[0].set_title("Länge der HRV-Tracks")
ax[1].hist(ir_track_lengths,bins=20)
ax[1].set_title("Länge der IR-Tracks")
```
