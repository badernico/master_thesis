```python
lcol = {0:'blue',1:'red'}
marker = {0:'o',1:'v'}
```

```python
fig,ax,m = radolan_rx_map(linecolour='black',figsize=(10,10))
plt.sca(ax)
for i, case in haci_cases.iterrows():
    l = case.l00 + ((case.l01 - case.l00)//2)
    c = case.c00 + ((case.c01 - case.c00)//2)
    x,y = m(r.lon[l,c], r.lat[l,c])
    ax.plot(x,y,color=lcol[case.lightning],marker=marker[case.lightning],markersize=2)
ax.text(0.04,0.95,"N no lightning = {}".format(len(haci_cases[haci_cases.lightning==0])),
        fontsize=13, va='bottom', color='blue',
        transform=ax.transAxes,bbox=dict(boxstyle="square",ec='k',fc='w')
     )
ax.text(0.04,0.90,r"N with lightning = {}".format(len(haci_cases[haci_cases.lightning==1])), 
        fontsize=13, va='bottom',
        color='red',transform=ax.transAxes,bbox=dict(boxstyle="square",ec='k',fc='w')
     )     
ax.set_title("Location of radar-based CI, 2012 to 2014")
plt.savefig("/vols/satellite/home/lenk/notebooks/Vortrge/Doktorandenvortrag3/HACI_Fallkarte_2012-2014_Blitz.pdf",bbox_inches='tight')
```

```python
r = np.zeros((25,25),dtype="int16")
g = np.zeros((25,25),dtype="int16")
b = np.zeros((25,25),dtype="int16")
```

```python
r[24,13] = 200
r[11,24] = 100
g[11,14] = 235
b[5,5] = 254

r[10,11] = 125
g[10,11] = 125
b[10,11] = 12
```

```python
r = np.ma.masked_equal(r,0)
g = np.ma.masked_equal(g,0)
b = np.ma.masked_equal(b,0)
```

```python
a = r.mask & g.mask & b.mask
```

```python
import fixed_colourbar as fc
```

```python
fig,ax = plt.subplots(1,4,figsize=(20,5))
rp  =ax[0].imshow(r,cmap='Reds',vmin=0,vmax=255)
ax[0].set_title("rot")
fc.colourbar(rp)

gp = ax[1].imshow(g,cmap='Greens',vmin=0,vmax=255)
ax[1].set_title("grn")
fc.colourbar(gp)

bp = ax[2].imshow(b,cmap='Blues',vmin=0,vmax=255)
ax[2].set_title(u"blau")
fc.colourbar(bp)

ap = ax[3].imshow(~a*1,cmap='gray',vmin=0,vmax=1)
ax[3].set_title(u"alpha")
fc.colourbar(ap)
```
