Die NetCDF4-Datei enthält mehrere Gruppen, in die die Felder gruppiert sind. Der Aufbau ist wie folgt:

1. Gruppe **channels**: enthält Daten der infraroten SEVIRI-Kanäle
    * hrv: HRV-Kanal-Reflektanz
    * wv062: WV062-Kanal-Helligkeitstemperatur
    * wv073: WV073-Kanal-Helligkeitstemperatur
    * ir087: IR087-Kanal-Helligkeitstemperatur
    * ir097: IR097-Kanal-Helligkeitstemperatur
    * ir108: IR108-Kanal-Helligkeitstemperatur
    * ir120: IR120-Kanal-Helligkeitstemperatur
    * ir134: IR134-Kanal-Helligkeitstemperatur
   
2. Gruppe **cloud_masks**: Wolkenmasken mit für konvektive Wolken interessanten Bereichen
    * berendes_mask: Convective Cloud Mask nach *Berendes et al.* (2008)
    * nwcsaf_mask: Wolkenmaske der NWCSAF-Satellitenprodukte
    
3. Gruppe **cloud_types**: Wolkentypen
    * berndes_ct: Wolkentypen der Convective Cloud Mask nach *Berendes et al.* (2008)
    * nwcsaf_ct: NWCSAF-Wolkentypen
    
4. Gruppe **cmsaf_products**: wolkenmikrophysikalische Satellitenprodukte der CMSAF-CPP-Software
    * cpp_cot: optische Wolkendicke
    * cpp_cph: Wolkenphase
    * cpp_cph_ir: aus Infrarotkanälen abgeleitete Wolkenphase
    * cpp_qa: Qualitätskennzeichnung der CPP-Produkte
    * cpp_reff: Effektivradius
    
5. Gruppe **coordinate_references**: Koordinatenreferenzen für die unterschiedlichen Auflösungen und Datenquellen
    * hrv_latitude: geographische Breiten der HRV-Pixel, 153x153 px
    * hrv_longitude: geographische Längen der HRV-Pixel, 153x153 px
    * latitude: geographische Breiten der Standardauflösung, 51x51 px
    * longitude: geographische Längen der Standardauflösung, 51x51 px
    * radar_latitude: geographische Breiten der Radarpixel, 153x153 px
    * radar_longitude: geographische Längen der Radarpixel, 153x153 px
    
6. Gruppe **interest_fields**: aus den SEVIRI-Kanälen angeleitete Mecikalski-Indikatorenfelder
    * b108: IR108-Helligkeitstemperatur 
    * b62_108: Helligkeitstemperaturdifferenz zwischen WV062 und IR108 
    * b62_73: Helligkeitstemperaturdifferenz zwischen WV062 und WV073
    * b87_108: Helligkeitstemperaturdifferenz zwischen IR087 und IR108
    * cool15: 15-Minutentendenz der IR108-Helligkeitstemperatur
    * cool30: 30-Minutentendenz der IR108-Helligkeitstemperatur
    * db62_73_t15: 15-Minutentendenz der Helligkeitstemperaturdifferenz zwischen WV062 und WV073
    * db62_73_t30: 30-Minutentendenz der Helligkeitstemperaturdifferenz zwischen WV062 und WV073
    * db87_108_t15: 15-Minutentendenz der Helligkeitstemperaturdifferenz zwischen IR087 und IR108
    * dtri_t15: 15-Minutentendenz der Dreikanalhelligkeitstemperaturdifferenz von IR087, IR108 und IR120
    * indicator_sum: Summe der Indikation für jedes Pixel
    
7. Gruppe **nwcsaf_products**: NWCSAF-Satellitenprodukte
    * ctth: Wolkenoberkantenhöhe in m
    
8. Gruppe **radar**: Radarrefliktivitätsfaktor des RX-RADOLAN-Produktes
    * radar: RADOLAN-RX-Produkt
   
9. Gruppe **time_reference**: Trackzeitpunkte
    * time: Zeit der Trackschritte in Sekunden seit dem 1.1.0001 00:00 UTC


Die Variable track ist ein Dictionary und enthält jetzt Satelliten- und Radardaten, sowie Satellitenprodukte für alle Zeitschritte:

```python
track.keys()
```

Die Felder stehen jetzt für weitere Analysen zur Verfügung und können, z.B., dargestellt werden:

```python
plt.rcParams["figure.figsize"] = [8,6]
plt.pcolormesh(track['longitude'][0],track['latitude'][0],track['cpp_cot'][0],cmap='spectral')
plt.colorbar()
```

Oder auch alle Zeitschritte hintereinander:

```python
cot_stapel = np.vstack(track['cpp_cot'])
```

```python
plt.rcParams["figure.figsize"] = [7,len(track['cpp_cot'])*7]
plt.imshow(cot_stapel,cmap='spectral',vmin=0,vmax=100,interpolation='none')
plt.colorbar()
```
