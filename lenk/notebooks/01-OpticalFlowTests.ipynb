## Getting the flow

For the optical flow, we need to get the previous or future sat.-image. We take the next, now, and re-initialize the sat.-scence.

```python
dt_minutes = 15

dt = datetime.timedelta(minutes = dt_minutes)

s = MSevi(time = t + dt, scan_type = scan_type)
s.load('IR_108')
s.rad2bt()
```

We make the same cutout now. We can use the geo-ref from before...

```python
bc2 = gi.cutout_field4box(s.bt['IR_108'], ind, 51)
```

```python
pl.pcolormesh(clon, clat, bc2, cmap = cmap, vmin = 210, vmax = 300)
pl.xlabel('longitude')
pl.ylabel('latitude')
```

First, to get a feeling for the advection, we plot the Eulerian difference in BT.

```python
pl.pcolormesh(clon, clat, bc2 - bc, cmap = pl.cm.RdBu_r, vmin = -10, vmax = 10)
pl.xlabel('longitude')
pl.ylabel('latitude')
```

Okay, we see that the anvil is expanding, leading to the significant cooling at the anvil edges.

Next, we apply an optical flow method to estimate the spatial shift.

* first, BT data are stacked in time
* second, all parameters are supplied to the optical flow method

```python
# time stack
b = np.array([bc, bc2])

u, v = oflow.flow_velocity(clon, clat, b, dt = dt_minutes, flow_input = None, vmin = 200., vmax = 300.)
```

The parameters of the optical flow method are as follows:
 * dt is the time step in minutes
 * flow_input is an pre-calculated index shift
 * vmin is the lower bound of the field
 * vmax is the upper bound of the field

```python
pl.hist(u.flatten(), 20)
pl.hist(v.flatten(), 20, alpha = 0.5)
pl.xlabel('velocity (m/s)')
```

At the end, we plot the wind on top of the BT field.

```python
xc, yc = gi.ll2xy(clon, clat)

pl.pcolormesh(xc, yc, bc, cmap = cmap, vmin = 210, vmax = 300)

ns = 4
qplot = pl.quiver(xc[::ns, ::ns], yc[::ns,::ns], u[0,::ns,::ns], v[0,::ns,::ns], color = 'yellow')
pl.xlabel('x-distance')
pl.ylabel('y-distance')
qk = pl.quiverkey(qplot, 0.5, 0.97, 10, r'$10 \frac{m}{s}$')
```

Okay, this looks at least reasonable.
