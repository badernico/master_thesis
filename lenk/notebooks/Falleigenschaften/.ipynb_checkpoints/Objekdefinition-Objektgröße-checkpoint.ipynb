In diesem Notebook sehen wir uns die satellitenbasierten Eigenschaften der HACI-Objekte zum Auslösungszeitpunkt in Abhängigkeit von den Parametern der Objektdefinition an.

```python
haci_data_files = sorted(haci_data_files)
```

```python
def get_object_values(data_array,object_mask):
    return data_array[np.where(object_mask==1)]
```

```python
def get_id_of_centre_object(labelled_field):
    from scipy import ndimage as ndi
    """
    Selects the object which contains the central point of the cutout in the 
    centre of all available field layers.
    
    Input:
        * labelled_field: array like, 2d
            fields to find the object in
    Returns:
        * wanted_id: int
            identifier of the wanted object
    """
    
    row_shape = labelled_field.shape[0]
    col_shape = labelled_field.shape[1]
    
    object_labels = np.unique(labelled_field)[1:]
    wanted_id = 0
     
    for o in object_labels:
        idx = np.where(labelled_field==o)
        
        if row_shape//2 in idx[0] and col_shape//2 in idx[1]:
            wanted_id = o
        else:
            wanted_id = 0
            
    # check if the wanted_id still is zero, if yes check the distances of object centroids
    if wanted_id==0:
        distances = dict()

        centroids = ndi.measurements.center_of_mass(labelled_field,
                                                    labelled_field,
                                                    object_labels)

        for i,c in enumerate(centroids):
            distances[object_labels[i]] = np.sqrt((row_shape//2-c[0])**2 + (col_shape//2-c[1])**2)

        # check for the minimum distance
        min_dist = np.min(np.array(list(distances.values())))

        if min_dist > 10:
            wanted_id = 0
        else:
            wanted_id = object_labels[np.argmin(np.array(list(distances.values())))]
        
    return wanted_id
```

## 1. Ein Beispiel

```python
test_file = fio.read_nc_file(haci_data_files[3668])
```

```python
plt.imshow(test_file['ir108'][6],vmin=210,vmax=300,cmap="gray_r")
```

```python
objects = np.array([oseg.watershed_local_min_segmentation(ir,6,220,273.15,240,5,0) for ir in test_file['ir108']])
```

```python
plt.imshow(test_file['ir108'][6],vmin=210,vmax=300,cmap="gray_r")
plt.imshow(np.ma.masked_less(objects[6],1))
```

Das sieht ganz vernünftig aus. Die Objekte beschränken sich hauptsächlich auf den "Kern" der Wolken. Wir bekommen drei Objekte, von denen das mittlere dasjenige ist, das zu der HACI-Detektion gehört. 

```python
object_mask = np.ma.masked_equal(objects[6],ot.get_wanted_object_id(objects)).mask*1
```

```python
plt.imshow(object_mask)
```

```python
object_properties = {k:[] for k in list(test_file.keys())[:-7]}
```

```python
for k in list(test_file.keys())[:-7]:
    if k not in ['hrv','rx']:
        object_properties[k] = get_object_values(test_file[k][6],object_mask)
    elif k in ['hrv']:
        object_properties[k] = get_object_values(test_file[k][6],object_mask.repeat(3,axis=0).repeat(3,axis=1))
    else:
        om = object_mask.repeat(3,axis=0).repeat(3,axis=1)
        obj_lon = test_file['hlon'][np.where(om==1)]
        obj_lat = test_file['hlat'][np.where(om==1)]
        
        idx = gi.get_index_kdtree(np.array([[obj_lat],[obj_lon]]),test_file['rlat'],test_file['rlon'])
        
        x = []
        y = []

        for i in idx:
            x.append(i[0][0])
            y.append(i[1][0])
            
        object_properties['rx'] = test_file['rx'][6][x,y]
```

```python
depth = np.arange(3,11)
tmin = np.arange(210,235,5)
tmax = np.arange(273.15,288.15,5)
tlevel = np.arange(220,255,5)
spread= np.arange(1,11)
```

```python
parameter_sets = {}
depths = []
tmins= []
tmaxs = []
tlevels = []
spreads = []
```

```python
i = 0
for d in depth:
    for tm in tmin:
        for tx in tmax:
            for tl in tlevel:
                for s in spread:
                    parameter_sets[i] = {'depth':d,'tmin':tm,'tmax':tx,'tlevel':tl,'spread':s}
                    depths.append(d)
                    tmins.append(tm)
                    tmaxs.append(tx)
                    tlevels.append(tl)
                    spreads.append(s)
                    i += 1
```

```python
object_masks = {}
```

```python
for k in parameter_sets.keys():
    object_masks[k] = oseg.watershed_local_min_segmentation(test_file['ir108'][6],
                                                            parameter_sets[k]['depth'],
                                                            parameter_sets[k]['tmin'],
                                                            parameter_sets[k]['tmax'],
                                                            parameter_sets[k]['tlevel'],
                                                            parameter_sets[k]['spread'],0)
```

```python
object_size = []
object_numbers = []

for k in object_masks.keys():
    mask = np.ma.masked_equal(object_masks[k],get_id_of_centre_object(object_masks[k])).mask*1
    object_size.append(len(np.where(mask==1)[0]))
    object_numbers.append(np.max(object_masks[k]))
```

## 1.1 Objektgröße und Anzahl von Objekten im Ausschnitt

```python
fig,ax = plt.subplots(7,1,figsize=(35,14),sharex=True)
ax[0].plot(depths)
ax[0].set_title("Tiefe des lokalen Minimums")
ax[0].set_ylabel(r"$T_\mathrm{B}$ / K")
ax[1].plot(tmins)
ax[1].set_title("Minimumtemperatur der Objektdefinition")
ax[1].set_ylabel(r"$T_\mathrm{B}$ / K")
ax[2].plot(tmaxs)
ax[2].set_title("Maximumtemperatur der Objektdefinition")
ax[2].set_ylabel(r"$T_\mathrm{B}$ / K")
ax[3].plot(tlevels)
ax[3].set_title("Konstanter Schwellwert der Objektdefinition")
ax[3].set_ylabel(r"$T_\mathrm{B}$ / K")
ax[4].plot(spreads)
ax[4].set_title("Aufspreizung der Objektdefinition")
ax[4].set_ylabel(r"$T_\mathrm{B}$ / K")
ax[5].plot(np.array(object_size)*21.744)
ax[5].set_title("Objektgröße")
ax[5].set_ylabel(u"A / km²")
ax[6].plot(object_numbers)
ax[6].set_title("Anzahl der Objekt im Ausschnitt")
plt.savefig("{}/2019-01_trackingstudie/pics/Parametervariation_Objektdefinition_Beispielfall.pdf".format(proj_path))
```

Die Objektgröße ist im Beispiel relativ unempfindlich gegenüber den Parametern der Objektdefinition. Bei einer Tiefe des lokalen Minimums von 4&thinsp;K gibt es allerdings einen Sprung und das Objekt wird ziemlich groß und nimmt ca. 80&thinsp;% des gesamten Ausschnitts ein. Ansonsten ist das zum HACI-Objekt gehörende Satellitenobjekt unabhängig von der Parameterwahl 4&thinsp;px<sup>2</sup> = 87&thinsp;km<sup>2</sup> groß. Dieser Sprung könnte auch daran liegen, wie die Objekte aus den lokalen Minima gebildet werden. Das sollte nochmal überprüft werden.

Die Anzahl der erkannten Objekt schwankt stärker, wird aber ab einer Tiefe des lokalen Minimums von mehr als 7&thinsp;K stabil bei drei Objekten. Wenn man sich den Ausschnitt betrachtet, ist das die wahrscheinlich sinnvollste Anzahl von Objekten.

Zusammengefasst scheint die Tiefe des lokalen Minimums sowohl für die Objektgröße als auch für die Objektanzahl der entscheidende Parameter zu sein. Dieser Parameter bestimmt, wieviele mögliche Objektkandidaten es gibt. Warum dieser Parameter auch für die Objektfläche ausschlaggebend ist, muss allerdings noch genauer untersucht werden.
