```python
def plot_haci_case(case_data):
    from plotting_tools.colormaps import enhanced_colormap2
    emap = enhanced_colormap2()
    
    from plotting_tools.colormaps import radar_cmap as rc
    
    fig,ax = plt.subplots(4,4,figsize=(20,20),sharex=True,sharey=True)
    axs = ax.ravel()

    for i,ir in enumerate(case_data['ir108']):
        p = axs[i].pcolormesh(case_data['slon'],case_data['slat'],ir,vmin=210,vmax=300,cmap=emap)
        p1 = axs[i].contourf(case_data['rlon'],
                             case_data['rlat'],
                             np.ma.masked_greater(np.ma.masked_less(case_data['rx'][i],5),70),
                             np.arange(5,75,5),
                             vmin=5,vmax=70,cmap=rc(),
                             alpha=0.5)
        
        axs[i].set_title("{}\n t = {} min".format(np.datetime_as_string(case_data['time'][i],
                                                                       unit='m',
                                                                       timezone='UTC'),
                                         (i*5) - 30))

    cbar1 = fig.colorbar(p1, ax=ax.flat,orientation='horizontal')
    cbar1.ax.set_title(u'RX / dBZ')

    cbar = fig.colorbar(p, ax=ax.flat)
    cbar.ax.set_title(r'$T_\mathrm{B}$(10.8 µm) / K')
    
    fig.delaxes(axs[13])
    fig.delaxes(axs[14])
    fig.delaxes(axs[15])
                         
    return fig 

def plot_case_object(case_data,object_data):
        
    fig,ax = plt.subplots(4,4,figsize=(20,20),sharex=True,sharey=True)
    axs = ax.ravel()

    for i,ir in enumerate(case_data['ir108']):
        p = axs[i].imshow(ir,vmin=210,vmax=300,cmap='gray_r')
        p1 = axs[i].imshow(np.ma.masked_less(object_data[i],1),
                           alpha=0.6,vmin=1,vmax=np.max(object_data),cmap='nipy_spectral')
        
        axs[i].set_title("{}\n t = {} min".format(np.datetime_as_string(case_data['time'][i],
                                                                        unit='m',
                                                                        timezone='UTC'),
                                                  (i*5) - 30)
                                                 )
    cbar1 = fig.colorbar(p1, ax=ax.flat,orientation='horizontal')
    cbar1.ax.set_title(u'Objektnummer')

    cbar = fig.colorbar(p, ax=ax.flat)
    cbar.ax.set_title(r'$T_\mathrm{B}$(10.8 µm) / K')
    
    fig.delaxes(axs[13])
    fig.delaxes(axs[14])
    fig.delaxes(axs[15])
                         
    return fig  
```

## 1.2 Auslösungszeitpunkt aller Fälle am Falltag

```python
vals = []
ts = []
fs = []
case_ids = []
object_arrays = dict()
```

```python
for case_id in case_data.keys():
    try:
        case_objects, case_tracks = ot.track_objects_lmin_cost(case_data[case_id],'ir108','wv073',
                                                               minimum_depth=4,
                                                               tmin=220,tmax=273.15,tlevel=240,
                                                               spread=5,search_factor=5,smoothing_factor=1)
        object_id = ot.get_wanted_object_id(case_objects)
        wanted_object_id = "{:03d}_{:04d}".format(len(case_objects)//2,object_id)
        wanted_track = ot.select_wanted_track_from_list(case_tracks,wanted_object_id)
        wanted_object = ot.get_wanted_object_field(case_objects,wanted_track)

        values = ot.collect_object_values(case,wanted_object)

        for field in values.field.unique():
            field_data = values[values.field==field]
            lfd = len(field_data.value)

            vals.extend(field_data.value.values)
            ts.extend(field_data.time.values)
            fs.extend(field_data.field.values)
            case_ids.extend([case_id]*lfd)

        object_arrays[case_id] = wanted_object
    except:
        continue
```

```python
case_object_values_df = pd.DataFrame({'field':fs,'value':vals,'time':ts,'case_id':case_ids})
```

Die meisten Objekte sind zum Auslösungszeitpunkt noch sehr warm. Nur zwei sind in einem Temperaturbereich, den man für konvektive Objekte erwarten könnte. Diese Objekte liegen entweder in einem Wolkenfeld das schwer zu trennen ist, sind sehr groß, die Objektdefinition scheitert hier oder sie fangen sehr zeitig im konvektiven Lebenszyklus an zu regnen.

```python
object_areas = dict()
object_area_tci = []
object_area_tci_case = []
for case in case_object_values_df.case_id.unique():
    a = []

    for time in case_object_values_df.time.unique():
        a.append(len(case_object_values_df[(case_object_values_df.time==time) & 
                                           (case_object_values_df.case_id==case) & 
                                           (case_object_values_df.field=='ir108')].values))
        if time == 0:
            object_area_tci.append(len(case_object_values_df[(case_object_values_df.time==time) & 
                                           (case_object_values_df.case_id==case) & 
                                           (case_object_values_df.field=='ir108')].values))
            object_area_tci_case.append(case)

    object_areas[case] = a
```

```python
object_area_tci_df = pd.DataFrame({'area':object_area_tci,'case_id':object_area_tci_case})

plt.figure(figsize=(16,8))

sns.catplot(x='case_id',
            y='area',
            data=object_area_tci_df,
            )
plt.xlabel(r"$\left(t - t_\mathrm{CI} \right)$ / min")
plt.ylabel(u"Objektfläche/ px²")
plt.xticks(rotation=90)
plt.title("Objektfläche zum Auslösungszeitpunkt")
plt.gcf().set_size_inches(16, 8)
```

Die meisten Fälle des ausgewählten Falltages zeigen keine für ein konvektives Objekt zu erwartende Entwicklung der Objektfläche. Bei vier Fällen ist die Objektfläche bereits 30&thinsp;min vor dem Auslösungszeitunkt mehr als 50&thinsp;px² groß. Bei drei dieser Fälle gibt es immer wieder sprunghafte Änderungen der Objektfläche. Hier könnte es Probleme mit der Objektdefinition geben. Einer dieser Fälle zeigt ein Schrumpfungssignal. Bei diesen Fällen handelt es sich wahrscheinlich nicht um wachsende Konvektion, sondern um bereits wieder zerfallende oder keine Konvektion.

Vier Fälle zeigen einen Wachstum der Objektfläche im betrachteten Zeitraum. Bei deisen vier Fällen handelt es sich um die Fälle, die zum Auslösungszeitpunkt mittlere Helligkeitstemperaturen im Bereich um 260&thinsp;K und maximale Helligkeitstemperaturen  von unter 280&thinsp;K haben oder zumindest Pixel mit solchen Werten beinhalten. Ihre Fläche ist zum Auslösungszeitpunkt im Bereich unter 50&thinsp;px<sup>2</sup>.

Daraus lassen sich erste Kriterien für eine satellitenbasierte Filterung der HACI-Objekte ableiten.

1. eine Zunahme der Objektfläche um den Auslösungszeitpunkt herum haben
2. eine Abnahme der mittleren und minimalen Helligkeitstemperatur um den Auslösungszeitpunkt herum haben
3. Die Objektfläche zum Auslösungszeitpunkt sollte kleiner als 50&thinsp;px<sup>2</sup> sein
4. Die maximale IR-10,8-µm-Helligkeitstemperatur des Objekts sollte kleiner als 280&thinsp;K sein
5. Die mittlere IR-10,8-µm-Helligkeitstemperatur des Objekts sollte kleiner als 260&thinsp;K sein

Inweit das typische Eigenschaften und damit auch gute Filterkriterien sind muss allerdings noch mit einer größeren Fallanzahl überprüft werden.
