# Blending Rank Fields

The seamless transition between observation-based nowcast and numerical simulation might involve a blending step in which both field are combined with a certain weight. The latter changes in time to reflect the information loss in the nowcast.

The blending might be split up across several components; STEPS is using different scales for this. However, an alternative approach might be to use structual and distributional characteristics separately. This would also allow for correcting distributional biases in the simulation.

Below, I try to explain how this might be implemented in python.

## Initializing Libraries

```python
%matplotlib inline

import os, sys
import numpy as np
import scipy.ndimage, scipy.interpolate
import pylab as pl
```

## Setting the test fields

The actual characteristics of the test field do not matter. 

I constuct a spatially correlated field for the nowcast.

```python
shape = (300,300)


# setting decorrelation scale and distribution
nowcast = 0

for scale in range(9):
    r = np.random.uniform(size = shape)
    nowcast +=  (2*scale)**2 * scipy.ndimage.gaussian_filter(r, 2**scale / 4. ) 

nc = 6 * 10**((nowcast  - nowcast.min())/(nowcast.max() - nowcast.min()))

pl.subplot(121)
pl.imshow(nc)

pl.subplot(122)
pl.hist(nc.flatten())
pl.xscale('log')
pl.yscale('log')
```

And the same for the simulation, but with different parameters and transformation.

```python

# setting decorrelation scale and distribution
simul = 0.

for scale in range(4, 9):
    r = np.random.uniform(size = shape)
    simul +=  (scale)**3 * scipy.ndimage.gaussian_filter(r, 2**scale / 2.) 

sim = 60 * ((simul  - simul.min())/(simul.max() - simul.min()))**4

pl.subplot(121)
pl.imshow(sim)

pl.subplot(122)
pl.hist(sim.flatten())
pl.hist(nc.flatten(), alpha = 0.5)
pl.xscale('log')
pl.yscale('log')
```

Structure and distribution of the test data are different. The idea is the analyze and blend them separately.

## Rank transformation

To get the structure of the field, we apply a rank transformation.

```python
######################################################################
######################################################################


def rank_transformation(f, normalize = True, gamma = 1.):

    '''
    The routine performs rank transformation of a field, meaning that 
    the ranks of the individual field values are returned.


    INPUT
    =====
    f: field
    normalize: optional, option if the rank field should be normalized to one
    gamma: optional, gamma correction factor


    OUTPUT
    ======
    f: field ranks
    '''


    # (i) make vector out of field matrix............................
    fvec = f.flatten()

    # (ii) sort the field ............................................
    fsort = np.sort( fvec )

    # (iii) calculate the ranks of the field .........................
    frank = np.searchsorted( fsort, fvec ).reshape( f.shape )


    # optional things ................................................
    if normalize:
        frank = (1. * frank ) / frank.max()

    if gamma != 1:
        frank = frank**gamma

    return frank

######################################################################
######################################################################

```

```python
nc_rank = rank_transformation(nc)
sim_rank = rank_transformation(sim)

pl.figure(figsize = (16,16))
pl.subplot(211)
pl.imshow(nc_rank)
pl.subplot(212)
pl.imshow(sim_rank);
```

## Combining rank fields

Now, the rank field might be combined. I make it simple here, however, using different weight for different spatial scales (like in STEPS) might be more appropriate.

```python
weight = 0.3

comb_rank = weight * nc_rank + (1 - weight) * sim_rank

pl.figure(figsize = (16,16))
pl.subplot(211)
pl.imshow(comb_rank);
```

Now, the combined rank field is not uniformly distributed anymore ...

```python
pl.hist(comb_rank.flatten());
```

Hence, we apply the rank transformation a second time ...

```python
crank = rank_transformation(comb_rank)

pl.figure(figsize = (16,16))
pl.subplot(211)
pl.imshow(crank)
```

## Back transformation and distribution blending

For the back transfromation, we need the inverse cummulative distribution function that assigns a certain threshold value to a cdf value.



```python
def inverse_cdf(f, hbins = 100):
    
    '''
    Function that calculates inverse cdf transform for a given field.
    
    INPUT
    =====
    f: 2d field
    
    
    OUTPUT
    ======
    icdf: function for inverse cdf transform
    '''
    
    # calculate histogram
    h, xe = np.histogram(f.flatten(), hbins)
    
    # get bin mid-points
    xmid = 0.5 * (xe[1:] + xe[:-1])
    
    # cummulative distribution function
    cdf = h.cumsum() / np.float(h.sum())
  
    # filter to ensure increasing cdf values
    cdf = scipy.ndimage.gaussian_filter1d(cdf, 0.5)
    
    # extend to full values range
    cdf = np.hstack([0,cdf,1])
    xm = np.hstack([f.min(), xmid, f.max()])
    
    # invsersion
    icdf =  scipy.interpolate.interp1d(cdf, xm)
    
    
    return icdf
```

We obtain the different distribution mappings

```python
icdf_nowcast = inverse_cdf(nc)
icdf_sim = inverse_cdf(sim)
```

We perform the distribution blending, perhaps taking a different weight.

```python
dweight = 0.6

comb = weight * icdf_nowcast(crank) + ( 1 - weight ) * icdf_sim(crank)

pl.figure(figsize = (16,16))
pl.subplot(211)
pl.imshow(comb)
```

Hence, the structural and distributional blending can be done with different time scales.
