```python
# plot tracks
fs = glob.glob("/Users/nicobader/Documents/Uni_Leipzig/master_thesis/analysis/tracks/data/"+y_oi+"/"+m_oi+"/"+d_oi+"/UTM_*")

plt.figure(figsize=(15,16))
ax = plt.axes(projection = ccrs.UTM(zone = 32))
ax.set_extent([4.708, 16.269 , 46.552, 55.691], crs=ccrs.PlateCarree())
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.3, linestyle='--')
gl.xlabels_top = False
gl.xlabel_style = {'size': 14}
gl.ylabel_style = {'size': 14}
gl.ylabels_right = False
ax.add_feature(cf.COASTLINE)
ax.add_feature(cf.BORDERS, linewidth = 1.2)
ax.add_feature(cf.STATES, linewidth = 0.1)
ax.text(ax.get_xlim()[0]+((ax.get_xlim()[1]-ax.get_xlim()[0])*0.115), ax.get_ylim()[1]-((ax.get_ylim()[1]-ax.get_ylim()[0])*0.04), doi, ha = 'center', fontweight = 'bold', fontsize=15,bbox=dict(facecolor='white', edgecolor='grey', pad=5.0, alpha = 0.7))

for i in np.arange(0,len(fs),1):
    # open track data
    track = pd.read_csv(fs[i])
    # transform to UTM coordinates (Zone 32)
    plt.plot(track.longitude, track.latitude, linewidth = 1.7, color = 'darkred')

plt.savefig('/Users/nicobader/Documents/Uni_Leipzig/master_thesis/analysis/tracks/output/plots/tracks_'+y_oi+m_oi+d_oi+'.png', dpi = 400, facecolor = 'white', transparency = False)
```
